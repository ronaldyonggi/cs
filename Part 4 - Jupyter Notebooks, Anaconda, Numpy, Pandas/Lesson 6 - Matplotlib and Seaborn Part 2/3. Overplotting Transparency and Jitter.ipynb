```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
fuel_econ = pd.read_csv('./data/fuel_econ.csv')
fuel_econ.head(2)
```

```python
# Analyze plot vs. combined fuel efficiency
sb.regplot(data = fuel_econ, x = 'year', y = 'comb');

# Observation: there are apparent overlap in data points! 
# it would be useful if we can jitter those
```

```python
sb.regplot(data = fuel_econ, x = 'year', y = 'comb', x_jitter = 0.3);

# Observation: Still hard to distinguish data points. Next step:
# add transparency
```

```python
# alpha sets the opaqueness of points
sb.regplot(data = fuel_econ, x = 'year', y = 'comb', x_jitter = 0.3,
          scatter_kws = {'alpha': 1/20});
```
