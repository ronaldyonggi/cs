**Task 2**: Let's look at the relationship between two other numeric variables. How does the engine size relate to a car's CO2 footprint? The 'displ' variable has the former (in liters), while the 'co2' variable has the latter (in grams per mile). Use a heat map to depict the data. How strong is this trend?

```python
# YOUR CODE HERE
max_displ = fuel_econ['displ'].max()
min_displ = fuel_econ['displ'].min()
max_co2 = fuel_econ['co2'].max()
min_co2 = fuel_econ['co2'].min()
bin_x = np.arange(min_displ, max_displ + 0.5, 0.5)
bin_y = np.arange(min_co2, max_co2 + 40, 40)
plt.hist2d(data = fuel_econ, x = 'displ', y = 'co2', bins = [bin_x, bin_y], cmin = 0.5, cmap = 'viridis_r');
plt.colorbar();

# Observation:
# 1. Apparent positive correlation: Larger engines correlate to greater CO2 footprint
```
