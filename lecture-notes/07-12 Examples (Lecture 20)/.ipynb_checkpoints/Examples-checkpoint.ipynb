# P-values  (Review)
## Hypothesis Testing Logic
* Define 2 mutually exclusive descriptions: either this or that
* One of them can be evaluated using probability (the null). Example:
    * A coin is fair so heads and tails should have the same probability
    * A die is fair, so each face should occur 1/6 of the time
* You can **reject the null** and accept the alternative
    * Otherwise, you're still unsure, but null looks reasonable
* Fisher (1935): "No isolated experiment, however significant in itself, can suffice for the experimental demonstration of any natural phenomenon"
    * In statistic, we're always dealing with uncertainty. Any time we take a sample, it's a random process .
    * We could be roll a die 20 times and all that 20 give out 6.

## Quantifying Conclusions
Steps:
0. Find some data. These are the **observations**
    * In this class, data are provided
    * In real life, we have to find the data by ourselves.
1. Come up with 2 descriptions of the world:
    * Null: Data come from a well-defined random process
    * Alternative: Something else is going on

We evaluate how unusual the data would be under the **null** Do the data supports the alternative hypothesis?
2. Choose a test statistic to summarize the data
    * Measurement of distance
    * The greater or lower this number, the stranger the result
3. Compute the following probability (p-value)

P value by definition: the test statistic would be **equal to or more extreme** than the observed test statistic under the null hypothesis.
* Draw simulations assuming that the null is true
* See how likely my result is to occur
* Draw conclusions from there
<img src = 'quantifying.jpg' width = 650 />
Recall the histogram above,
* The vertical blue line is what Mendel observed (observed test statistic)

**Other than the null hypothesis, how likely to get a result that is more strange (more towards the tail)?**
The area of the bars to the right side of Mendel's observed test statistic is the **p-value**!

## Conventions of Consistency
* **Inconsistent**: The test statistic is "in the tail" of the null distribution
* **"In the tail" first convention**
    * The area in the tail (p-value) is less than 5%
    * The result is "statistically significant"
* **"In the tail" second convention** -> More precise
    * The area in the tail (p-value) is less than 1%
    * The result is "highly statistically significant"
    
However, this is just convention. You can determine consistency arbitrarily.

## Choosing a Test Statistic
**Test statistic**: The statistic that you have chosen to calculate, to help you decide between the 2 hypotheses

**Goal**: If the null hypothesis is false, then you expect that measuring the test statistic will allow you to reject the null

## Coolness: Pick a Number between 1 - 100.
The number we picked would be the we coolness level. Let'say a whole class did this, and the instructor too.

We can ask a few questions:

#### Is the instructor is cooler than the class?
* Let's say instructor picked 85
* The test statistic is the instructor's coolness level
    * The p-value is everyone in the class who picked 85 or above

Let's say there were 50 students in the classroom and there're 2 students that picked 85 or above. 
* If we picked randomly one of the students in the class, there'll be 4% chance that we get the student that has coolness level of 85 or above
    * Our p-value would be 4% or 0.04
    * You can say the instructor is cooler than the class

#### Is the instructor less cool than the class?
* Let's say instructor picked 5
* Let's say out of 50 students, 4 students got coolness level 5 and below

This means if we pick a student from the class, there's a 8% chance that we'll get a coolness level of 5 or below. In this case, the test statistic is the same.

#### Is the instructor just different or weird?
In this case, what we want to know: How far away is the instructor's coolness level from the average coolness of the students in the class.
Let's say the students picked randomly from 1 - 100 and the average is 50.
If the instructor picks:
* 85, then the instructor is 35 away from the average
* 5, then the instructor is 45 away from the average

In this case, we want to measure how far we are from the middle (not how far we are in the positive or negative direction). This is why we take the absolute value.
* Take your coolness level
* Subtract it with 50
* Take the absolute value

**For the first 2 questions, the test statistic is the same. For the 3rd question, the test statistic is different!**


## Back to the Mendel's Pea Flowers

```python
null = make_array('Purple', 'Purple', 'Purple', 'White')

def test_stat(sample):
    proportion_purple = np.count_nonzero(sample == 'Purple') / len(sample)
    return abs(proportion_purple - 0.75)

# Mendel's experiment result below:
total_plants = 929
purple = 705
observed_statistic = abs(purple / total_plants - 0.75) # 705 were Purple
observed_statistic
```

```python
# Simulating the test statistic under the null hypothesis
repetitions = 5000
sampled_stats = make_array()

# Below is the old way. 
# for i in np.arange(repetitions):
#     new_sample = np.random.choice(null, total_plants)
#     sampled_stats = np.append(sampled_stats, test_stat(new_sample))

# Below is the more efficient way (using sample_proportions)
model_proportions = [0.75, 0.25]
def test_stat(proportions_array):
    return abs(sample_proportions(929, proportions_array).item(0) - 0.75)

for i in np.arange(repetitions):
    sampled_stats = np.append(sampled_stats, test_stat(model_proportions))
```

```python
null_dist = Table().with_column('Distribution of the test statistic under the null hypothesis', sampled_stats)
null_dist.hist()
_ = plt.plot([observed_statistic, observed_statistic], [0, 60])
```

We calculated the p-value:

```python
# The P-value (an approximation based on the simulation)
np.count_nonzero(sampled_stats >= observed_statistic)/repetitions
```

And we concluded that Mendel's result is likely to occur. 

### Discussion Question: What happens if we don't use the absolute value when calculating the test statistic?

```python
repetitions = 5000
sampled_stats = make_array()

model_proportions = [0.75, 0.25]
def test_stat(proportions_array):
    return sample_proportions(929, proportions_array).item(0) - 0.75

for i in np.arange(repetitions):
    sampled_stats = np.append(sampled_stats, test_stat(model_proportions))
```

```python
null_dist = Table().with_columns('Null distribution', sampled_stats)
null_dist.hist()
_ = plt.plot([observed_statistic, observed_statistic], [0, 60])
```

**The distribution and the p-value changes!**

```python
np.count_nonzero(sampled_stats >= observed_statistic) / len(sampled_stats)
# the previous p-value was around 0.54. Why is it different this time?
```

Going back to the coolness example. In the first 2 questions:
1. How many people got the instructor's coolness level or above?
2. How many people got the instructor's coolness level or below?

We didn't take the absolute value at all. But on the last question:
3. How many people were further from the average than the instructor?

We took the absolute value!

Now back to the pea-plant. Looking at the histogram:
* Negative values mean less than 75% purple flowers
* 0 means we got exactly 75% purple flowers
* Positive values mean greater than 75% purple flowers

1. When we use absolute value, we're calculating the distance away from the 75%. 
2. When we took away the absolute value, we're analyzing how likely we'll get 75% or above
    * This way we'll only taking the right side of the histogram, and thus we'll get lower area (lower p-value)
    
We also changed our alternative hypothesis. Instead of saying that the value should be far away from 75%, we're saying that the value should be far greater than 75%.

# Absolute Values & Alternatives
Our alternative can have a direction of association (+ or -)
* Coin example:
    * The coin can be biased towards heads
    * It could be biased towards less heads
* Smoking example:
    * Smoking has some relation to health (use absolute value)
    * Smoking is associated with poor health (no absolute value)

When our alternative hypothesis is a negative association, "equal to or more extreme" is a tail in the negative direction.

Depending on the question we want to ask:
* Choose the test statistic
* Choose a direction 

# Clarification Points
* The null and alternative hypotheses don't need to be complements (i.e. complete opposite) of each other
* You should not run multiple hypothesis tests on the same data
    * Isn't it convenient if we can just take an absolute value, then change the test statistic? NO! 
    * This makes the p-value invalid
* Rejecting the null doesn't always mean the alternative is true
    * Remember that the p-value is the probability of making a bad rejection
