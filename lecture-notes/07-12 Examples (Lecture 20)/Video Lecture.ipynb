# P-values  (Review)

## Hypothesis Testing Logic
In hypothesis testing, we define 2 mutually exclusive descriptions: either `this` or `that`. One of them is required to be evaluated using probability. We require the null hypothesis to define the world in terms of probability:

This "event" has some probability of occuring.
1. Our coin is fair, so heads and tails should occur equally half of the time
2. Our die is fair, so each face should occur 1/6 of the time

These are examples of valid `null` hypothesis since they specify a complete probability distribution.

If we decide to **reject the null**, then we accept the `alternative`. Otherwise, we **"fail to reject the null"**. The `null` still looks reasonable.

Recall Ronald Fisher (1935): "No isolated experiment, however significant in itself, can suffice for the experimental demonstration of any natural phenomenon".

Basically in statistic, we're always dealing with uncertainty. Any time we take a sample, it's a random process.

We could be unlucky, roll a die 20 times and all 20 give out 6s. 

We can never make a hard and fast conclusion when we do hypothesis test. But hypothesis test can give us a direction to look (i.e. the actual panel looks suspicious, we should look more into it).

## Quantifying Conclusions
In terms of conducting hypothesis test, 

Step 0 : Go find some data. These are what we call the `observations`
* Most of the time in the class, the data is provided. In the real world, we have to go out and get data by ourselves

Step 1 : Come up with 2 descriptions of the world:
1. `Null`: The data came from a well-defined random process
2. `Alternative`: Something else is going on

Evaluate ow unusual the data would be under the null.
* We went out to the world, grab some data. Do our data support the null hypothesis or is something else going on?
* Assume that the `null` is true, look at how the world behave by drawing simulations. See how likely our result would occur

Step 2: Choose a test statistic to summarize the data
* Choose a measurement of weirdness. This measurement would imply that the greater the value, the more weird the result.

Step 3: Draw simulations that compute the `P-value`.

By definition, `P-value` is the probability that the `test statistic` would be equal to or more extreme than the observed test statistic under the null hypothesis. 

Above is a long definition, but just remember:
1. Draw simulation
2. Assume that `null` is true
3. See how likely our result would occur
4. Then make conclusion

## Conventions of Consistency
When we discussed about `P-value`, we discussed about convention. 

We say that our data is `inconsistent` with our null hypothesis when our `test statistic` is **in the tail** of the null distribution.

By convention, we usually define **in the tail** as a `P-value` of less than 5%. We call this result **statistically significant**.

Sometimes we want a more precise cutoff, less than 1%. The result is **highly statistically significant**.

These are just convention. We can define consistency arbitrarily, whichever more convincing. 

* **Inconsistent**: The test statistic is "in the tail" of the null distribution
* **"In the tail" first convention**
    * The area in the tail (p-value) is less than 5%
    * The result is "statistically significant"
* **"In the tail" second convention** -> More precise
    * The area in the tail (p-value) is less than 1%
    * The result is "highly statistically significant"
    
However, this is just convention. You can determine consistency arbitrarily.

# Test Statistic

## Choosing a Test Statistic
`Test statistic`: The statistic that we have chosen to calculate, to help us **decide between the 2 hypotheses**

When we choose a `test statistic`, we have many options. For example, with the pea plants case, we did "absolute value - 0.75". Then we made a situation of what would happen if we removed the absolute value. 

Basically, the criteria for choosing a statistic is: We want to be able to say, "If the `null` hypothesis is false, then measuring the `test statistic` would help reject the `null`."

The further we are from the value that we expect, the more different our test statistic should be. An example would be the following activity: **Pick a number between 1 to 100**.

# Coolness - Pick a number between 1 to 100

Pick a number between 1 to 100. That would be our coolness level. 

We can ask a few questions:

#### Is the instructor is cooler than the class?

Let's say instructor picked 85. We ask the students' in the classroom for the coolness level that they chose. 

The question to ask: **"How many people got 85 or above?"**

The `test statistic` in this case is the instructor's coolness level. Students who picked 85 or above are the `P-value`. 

We try to find the probability of the instructor obtaining his coolness level from the population, the students in the classroom. 

We can calculate the probability by looking at how many of the students picked 85 or above. 

1. If the instructor was picked from the classroom (50 students),
2. and there were 2 students that picked 85 or above, then 
3. There will be about 4% chance of obtaining a coolness level of at least 85. Thus, the chance of the instructor having a coolness level of 85 is 4%.
4. The `P-value` is 4%, or 0.04.
5. The instructor is indeed cooler than the class

#### Is the instructor less cool than the class?

Let's say the instructor picked 5. "How likely is it to obtain coolness level of 5 or below?"

The question to ask: **"How many people got 5 or below?"**

There are 4 students in the classroom that picked 5 and below. 

If we picked a student at random from the class, there's a 4/50, or 8% chance that we will obtain coolness level of 5 or below.

The `test statistic` is the same as the previous case, **the instructor's coolness level**. We did not do anything to instructor's coolness level. All we did was to compare that value to those of the students'.

#### Is the instructor just different?
Is the instructor just unique? different?

The question to ask: **"How far away is the instructor's coolness level from the average coolness taken from the students in the classroom?"**

Supposed the students picked coolness level randomly from 1 to 100, and the average turned out to be 50. 

And let's say the instructor's coolness level is 85. 

"How different is the instructor from the average of the class"?

If we compute the difference between the instructor's coolness level, 85, and the classroom's average coolness level, 50, then the instructor is 35 away from the average.

If the instructor picked coolness level of 5, the instructor is now 45 away from the class average.

In this case, we want to measure how far our value is from the middle, not how far away the value is in the positive `(+)` or negative `(-)` direction. This is the case when we want to use **absolute value**. 

1. Take the instructor's coolness level
2. Subtract it from the class average of 50
3. Take the absolute value
4. That gives us the distance away from 50

# So what about the pea plants?
Below is the same setup as the previous lecture. We have the `null` distribution,

```python
null = make_array('Purple', 'Purple', 'Purple', 'White') # This is the null distribution
```

We have the `test_stat` function that counts the `test statistic` of a taken sample,

```python
def test_stat(sample):
    result = abs(np.count_nonzero(sample == 'Purple') 
                      / len(sample) 
                      - 0.75)
    return result
```

And Mendel's observation,

```python
# Mendel's experiment result below:
total_plants = 929
purple = 705
observed_statistic = abs(purple / total_plants - 0.75) # 705 were Purple
observed_statistic
```

We run the simulation below. 

```python
# Simulating the test statistic under the null hypothesis
repetitions = 5000
sample_stats = make_array() # An array that contains the resulting test statistics from samples

for i in np.arange(repetitions):
    # Calculate the test statistic of one sample using 'test_stat' function
    test_statistic = test_stat(np.random.choice(null, total_plants))
    sample_stats = np.append(sample_stats, test_statistic)
    
sample_stats
```

# Absolute Values & Alternatives
Our alternative can have a direction of association `(+)` or `(-)`

When we discussed about coin flip example, we say "the coin can be biased towards heads", or "the coin is biased towards less heads". In this case, the direction of association differs. We could be different in the `(+)` direction or `(-)` direction, or both. 

For example, smoking.
1. "Smoking has some relation to health" **[use absolute value]**
    * Which means smoking has some effect on health
    * We use absolute value in the `test statistic` to say, "smoking could make health better" or "could make health worse"

2. "Smoking is associated with poor health" **[don't use absolute value]**
    * Smoking makes health worse
    * Then we say, "Just look in the negative direction of some measurement of health".

    * Smoking has some relation to health (use absolute value)
    * Smoking is associated with poor health (no absolute value)
    
In the coolness example, we illustrated both of the cases above. 

**Is the instructor cooler than the rest of the class?** : Look at the positive coolness direction

**Is the instructor less cool than the rest of the class?** : Look at the negative coolness direction

**Is the instructor just different?** : Take the absolute value, then look at the positive direction (there is no negative in absolute values) 

When our alternative hypothesis is a negative association, "equal to or more extreme" is a tail in the negative direction.

It all depends on the `alternative` hypothesis. Depending on the question we want to ask:
* Choose the `test statistic`, then
* Choose a direction we want to look when we draw our `null` distribution

# Clarification Points
Some things to clarify,

### The null and alternative hypotheses don't need to be complements (i.e. complete opposite) of each other.

If the `null hypothesis` is that the coin is fair, does the `alternative` has to be that the coin is unfair? **NO**

The `null` and the `alternative` does not have to be the complete opposite of each other. We can have the "is the coin biased towards heads?" as the `alternative`. However, one has to be true and the other has to be not true.

It could be the case that the `null` is, "the coin is fair", while the `alternative` is, "the chance of getting heads is 0.7".. However, the probability of getting heads can be 0.3, 0.6, 0.9, etc. There are more possibilities that we did not consider in the hypothesis test. On the same time, the goal of the hypothesis test is just to decide between 2 scenarios.

If we test between 2 hypothesis and we reject the `null`, we don't necessarily accept the alternative. In the scenario above, we don't know if the chance of getting heads is 0.7. We just say that based on the 2 hypothesis, one of them seems more likely.
We don't know if any of the 2 is actually true, we only have a data that supports either one or the other. 

Even with the above, our conclusion is not always true. Recall we also have the `P-value` as a **type 2 error**, or the probability that we reject the `null` even when the `null` is correct. There is always uncertainty when we do hypothesis test.

### You should not run multiple hypothesis tests on the same data
Can we just take the absolute value when we change the `test statistic` to get a result that's statistically significant? **NO**! This changes the definition of the `P-value`. 

### Rejecting the null doesn't always mean the alternative is true
Recall that `P-value` is the probability of making a **bad rejection**. 

There is always some uncertainty when we make hypothesis test.
