# Classification
In the previous lecture, we discussed about prediction. Given data, can we predict about something that we have never seen before?

We have seen an example of prediction: regression. In regression, given a numerical data, we can use linear regression to predict some other numerical data. For example, we can use linear regression to predict the price of apartment in an area in the future given the number of people living in the area for the past years. 

Now we are going to discuss about another prediction problem that is fundamental to the machine learning field: **classification**. Classification is similar to the regression in a way that given some data, we make predictions. The difference is that that the data involved in classification is categorical, not numerical. 

## Classification Examples
<img src = 'spam.jpg' width = 300/>
Most emails has a convenient feature that **separates spams from the inbox**. In the past when this feature has not been implemented, most inboxes might be flooded by spams!

In this case, the classification is: "given an email containing texts, is the email a spam?"

Another classification problem is **self-driving cars**. Cameras are installed in cars, and given a video or an image captured by the camera, the car executes an action. For example, if the camera sees another car in front, then don't hit the car. In this case, the input of the classification problem is the image, the output is the stuff within the image (does the image contain a car? a person? etc)

<img src = 'scan.jpg' width = 400/>
Another classification is **medical scan determination**. Given a scan (e.g. x-ray) of a patient, does the patient have some kind of desease? This is especially common in cancer, where we can look at the shape of the tumor. 

Some other classification examples include predicting people's behavior for election, bank fraud detection and online dating. 

Classification is something that we do in daily life as well. For example, when we do grocery shopping for fruits, we determine that if a mango is squishy, then it should taste good. If the mango is hard, then it probably is not as good. With the mango example, there are attributes about some particular item and we used the combination of these attributes to make a judgment of whether to buy the mango. 

All these problem share the same element: we have an attribute about an input data and we try to make prediction or classification based on the data.

Determining whether a money bill is fake is another problem that can be solved using data science. Below we have a dataset of `banknotes` that contains image processing technique attributes that we will not discuss deeply. These attributes help us decide whether a bill is fake. For the `Class` variable, `1` indicates fake bill, while `0` indicates true bill.

```python
banknotes = Table.read_table('banknote.csv')
banknotes
```

And below we have the `WaveletVar` vs. `WaveletCurt` plot,

```python
banknotes.scatter('WaveletVar', 'WaveletCurt', colors = 'Class')
```

It might be difficult to make a judgment call based on the plot above since there are regions where the yellow dots and the blue dots overlap. How about we try `WaveletSkew` vs. `Entropy`?

```python
banknotes.scatter('WaveletSkew', 'Entropy', colors = 'Class')
```

It might be not easy to make a judgment call based on the plot above either.  

Below, we make a 3D plot of of `WaveletSkew`, `WaveletVar` and `WaveletCurt`

```python
color_table = Table().with_columns(
    'Class', make_array(1, 0),
    'Color', make_array('darkblue', 'gold')
)
color_table
```

First, we set so that the fake bill, `1`, is indicated by the darkblue color, while the real bill, `0`, is indicated by the gold color.

```python
banknotes = banknotes.join('Class', color_table)
banknotes
```

```python
ax = plt.figure(figsize=(8,8)).add_subplot(111, projection='3d')
ax.scatter(banknotes.column('WaveletSkew'), 
           banknotes.column('WaveletVar'), 
           banknotes.column('WaveletCurt'), 
           c=banknotes.column('Color'));
```

When we compare only `2` variables, we don't see a clear separation between fake bills and real bills. If we compare the `3` variables in a 3D plot like above, we can see the distinguishing feature between the `2` classes clearly!

The problem with what we've been describing so far: the way we have been distinguishing the 2 classes are very specific to the problem. In the CKD problem,

```python
ckd.scatter('Hemoglobin', 'Glucose', colors = 'Class')
```

From this particular problem, if an individual has `Hemoglobin` level greater than `13` and the `Glucose` level below `150`, then we can say that the individual shouldn't have CKD.

Then when we moved on to the fake bills problem,

```python
banknotes.show(3)
```

We had to go look at the data, plot the data and make judgment calls again. We had to come up with a rule that determines the distinguishing attributes between the 2 classes. Coming up with a rule by ourselves means that the rule would be very specific to this particular problem. 

In classification, we might go through a lot of datasets, a lot of classification problems, and come up with rules for distinguishing categorical datas. However, designing the rules is not easy. If we go back to the 3D model that we constructed,

```python
ax = plt.figure(figsize=(8,8)).add_subplot(111, projection='3d')
ax.scatter(banknotes.column('WaveletSkew'), 
           banknotes.column('WaveletVar'), 
           banknotes.column('WaveletCurt'), 
           c=banknotes.column('Color'));
```

Even if in 3D, the 2 classes might still be difficult to separate. It might be easier if we add a 4th variable, but 4D is unrealistic!

It appears that there are some properties in common between the 2 problems.

```python
ckd
```

In both cases, there are some attributes (e.g. `Age`, `Blood Pressure`, `WaveletVar`). The attributes themselves don't matter as long as those attributes let us distinguish between the 2 different classes.

When we notice the particular pattern in the problem, we might think that there is a way to write some computer program without us having to analyze the data, make scatter plots and come up with distinguishing rules. We might think that there is an algorithm that automatically figures out which variables separate the 2 data points. It turns out we can do that since there are a lot of properties in common with both problems. 

The 2 problems might seem different (one is CKD, one is fake bills), but there is a way to take the common properties of both problems. We can write a computer program that solve both of these problems!

# Classifiers
We have seen 2 different types of classification problems. If we look at each dataset and its data and variables, we can come up with rules on how to classify the points. 

Now, the goal is to write programs that work for general classification problems, not just the 2 that we saw. These programs are called `Classifiers`.

## Training a Classifier
<img src = 'classifier.jpg' width = 500/>
The way classifier works is:

1. Look at what's common to all the problems that we have observed before
    * All the problems have some attributes
        * If we shop for mangos, the attributes are: shape, size, softness
2. There is some classifer that if we give in attributes, it can predict whether the mango is tasty

<img src = 'classifier_2.jpg' width = 500/>
Classification ties back to inference in the following way:
<img src = 'population.jpg' width = 100/>

1. We have a population of all possible mangoes.

<img src = 'sample.jpg' width = 250/>

2. When we shop at supermarket, we only see a subset / sample of mangoes
    * Can we use the sample data to infer something about what we like about mangoes in the population? 
    * Thus, we come up with labels of "whether we like the mangoes or not"
    
<img src = 'labels.jpg' width = 300/>

3. Separate data to `Training set` and `Test set`. 

<img src = 'classifier_2.jpg' width = 500/>

Suppose we observe 100 mangoes, but pretend we only observe 70 of them. Based on the 70, we will figure out which attributes are the most useful to determine whether the mango is tasty. The `training set` means we are going to train ourselves to figure out how how to eat or like mangoes. 

We don't use the mangoes to decide which parts of the mango we like the most. Instead, we use the `test set` to evaluate accuracy.
