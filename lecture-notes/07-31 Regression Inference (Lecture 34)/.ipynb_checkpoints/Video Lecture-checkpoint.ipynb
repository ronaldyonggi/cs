# Regression Model
The past lecture, we discussed about a model for regression. We had a cloud of data points that appear to follow a linear relationship between `x` and `y`. We treat the data as if there is a real line that the data points follow, and that the data points are not perfect on the line because of errors, variation of random sampling, etc.

## A "Model": Signal + Noise
A model implies that there is a real line that exhibits relationship between `x` and `y`. "if we gain 10 pounds, or babies will gain 10 pounds". 
<img src = 'model.jpg' width = 500/>

The points don't fall exactly on the line, instead we say that the points fall according to "normal distribution with `mean` `0`". This means we expect half of the points fall above the real line, and the other half fall below the real line. 

The important thing is that we think that there is an actual line, and thus we can use a regression line. If instead we think the data follows a curve, then we won't be able to use a linear regression. 

## What We Get to See
In actuality, what we get to see is the following,
<img src = 'see.jpg' width = 400/>

We don't get to see the "real" lines. All we see is the points that are generated from that line. If we draw a regression line from the points, how close is the regression line from the actual line?

Below we have functions that are useful from previous lectures,

```python
def standard_units(any_numbers):
    """ Convert an array of numbers to standard units"""
    # (Value - mean) / SD
    return (any_numbers - np.mean(any_numbers)) / np.std(any_numbers)

def correlation(t, x, y):
    """ Return the correlation coefficient (r) of 2 variables. """
    # Average of the product of the `x` and `y`
    return np.mean(standard_units(t.column(x)) * standard_units(t.column(y)))

def slope(t, x, y):
    """ The slope of the regression line (original units). """
    r = correlation (t, x, y)
    # Slope = r * (SD of y / SD of x)
    return r * np.std(t.column(y)) / np.std(t.column(x))

def intercept(t, x, y):
    """ The intercept of the regression line (original units)"""
    # Average(y) - slope * Average(x)
    return np.mean(t.column(y)) - slope(t, x, y) * np.mean(t.column(x))

def fit(table, x, y):
    """ Return the fitted y value"""
    a = slope(table, x, y)
    b = intercept(table, x, y)
    return a * table.column(x) + b

def plot_residuals(t, x, y):
    """ Plot a scatter diagram and residuals. """
    t.scatter(x, y, fit_line =True)
    actual = t.column(y) # The actual y values
    fitted = fit(t, x, y) # The predicted y values
    residuals = actual - fitted
    print('r:', correlation(t, x, y)) # Print the r
    print('RMSE:', np.mean(residuals ** 2) ** 0.5) # Print the RMSE
    t.select(x).with_column('Residual', residuals).scatter(0, 1) # Plot the residual plot
    
```

```python
def standard_units(any_numbers):
    "Convert any array of numbers to standard units."
    return (any_numbers - np.mean(any_numbers)) / np.std(any_numbers)  

def correlation(t, x, y):
    """Return the correlation coefficient (r) of two variables."""
    return np.mean(standard_units(t.column(x)) * standard_units(t.column(y)))

def slope(t, x, y):
    """The slope of ther regression line (original units)."""
    r = correlation(t, x, y)
    return r * np.std(t.column(y)) / np.std(t.column(x))

def intercept(t, x, y):
    """The intercept of the regression line (original units)."""
    return np.mean(t.column(y)) - slope(t, x, y) * np.mean(t.column(x))

def fit(table, x, y):
    """Return the height of the regression line at each x value."""
    a = slope(table, x, y)
    b = intercept(table, x, y)
    return a * table.column(x) + b

def plot_residuals(t, x, y):
    """Plot a scatter diagram and residuals."""
    t.scatter(x, y, fit_line=True)
    actual = t.column(y)
    fitted = fit(t, x, y)
    residuals = actual - fitted
    print('r:', correlation(t, x, y))
    print('RMSE:', np.mean(residuals**2)**0.5)
    t.select(x).with_column('Residual', residuals).scatter(0, 1)
```

And below we have the plots of data points, actual and residuals

```python
def draw_and_compare(sample_size, true_slope=2, true_int=-5):
    x = np.random.normal(50, 5, sample_size)
    xlims = np.array([np.min(x), np.max(x)])
    errors = np.random.normal(0, 6, sample_size)
    y = (true_slope * x + true_int) + errors
    sample = Table().with_columns('x', x, 'y', y)

    sample.scatter(0, 1)
    plt.plot(xlims, true_slope*xlims + true_int, lw=2, color='green')
    plt.title('True Line, and Points Created')
    plt.show()

    sample.scatter(0, 1)
    plt.title('What We Get to See')
    plt.show()

    sample.scatter(0, 1, fit_line=True)
    plt.title('Regression Line: Estimate of True Line')
    plt.show()

    sample.scatter(0, 1, fit_line=True)
    plt.plot(xlims, true_slope*xlims + true_int, lw=2, color='green')
    plt.title("Regression Line and True Line")
    plt.show()
    
_ = interact(draw_and_compare, sample_size=make_array(10, 100, 400))
```

Above, as we increase the sample size, the regression line (blue line) will become closer to the true line (green line).

# The True Slope
How close is the regression line slope to the actual line slope?

## Confidence Interval for True Slope
Suppose we have a population and we are trying to estimate the parameter (e.g. `mean` baby weights). 

We take a sample, look at the `mean` baby weight of the sample


1. Bootstrap the scatter plot
    * We have data points. We can just select at random from those data points again with replacement. This will give us variation to the scatter plot
    
    
2. Find the slope of the regression line through the bootstrapped plot
    * Take a sample
    * Bootstrap them to obtain a resample 
    * Find the regression line from the resample 


Again, we assume that:
* The original sample is similar to the population
* The resamples look like the original sample


3. Repeat to obtain a bunch of different slopes


4. Draw the empirical histogram of all the generated slopes
    * Make sure the histogram of the generated slopes is roughly bell shaped so that it confirms that it meets the bootstrap criteria


5. Obtain the middle 95% interval

From the steps above, we can obtain an approximate 95% confidence interval for the slope of the true line. It gives us **a range of possible slopes that the actual slope might be in**. 

## Slope Inference
Below we have the **baby weights** dataset that we had in the past,

```python
baby = Table.read_table('baby.csv')
baby.show(3)
```

We are going to analyze the relationship between `Gestational Days` (how many days the baby was in the womb) and `Birth Weight`, in ounces. 

```python
plot_residuals(baby, 'Gestational Days', 'Birth Weight')
```

At a glance, the data points look good for a linear regression. Let's find the slope!

```python
slope(baby, 'Gestational Days', 'Birth Weight')
```

The slope above means if the baby is in the womb for 10 extra days, the baby would weight 4.6 ounces more.

If we try to take samples randomly from the baby 4 times and construct scatter plot out of them, we'll see the following,

```python
for i in np.arange(4):
    baby.sample().scatter(1, 0, fit_line=True)
```

Double check if the population and the sample have the same number of data points,

```python
baby.num_rows
```

```python
baby.sample().num_rows
```

Now we can do multiple resamples and construct the histogram!

```python
slopes = [] # An array that contains all the generated slope
for i in np.arange(5000):
    resample = baby.sample()
    resample_slope = slope(resample, 1, 0)
    slopes.append(resample_slope)
    
Table().with_column('Bootstrap Slopes', slopes).hist(bins = 20)
```

This means that if we increase the gestational days by 1, we expect the baby weights to increase by 0.35 to 0.60 ounces. 

If we calculate the 95% interval,

```python
left = percentile(2.5, slopes)
right = percentile(97.5, slopes)
[left, right]
```

Then we are 95% confident that the real population slope lies between the 2 numbers above.

Below is a function that combines the steps that we did above,

```python
def bootstrap_slope(table, x, y, repetitions = 5000):
    """ 
    table: a table that contains a sample (not population!)
    x and y: labels of data column that we want to analyze
    """
    
    # Bootstrap Resampling
    slopes = []
    for i in np.arange(5000):
        resample = table.sample()
        slopes.append(slope(resample, x, y))
        
    # Find the endpoints of the 95% confidence interval for the true slope
    left = percentile(2.5, slopes)
    right = percentile(97.5, slopes)
    
    # Slope of the regression line from the original sample
    observed_slope = slope(table, x, y)
    
    # Display results
    Table().with_column('Bootstrap Slopes', slopes).hist(bins = 20)
    plt.plot([left, right], [0, 0], color = 'yellow', lw = 8);
    print('Slope of regression line:')
    print(observed_slope)
    print('')
    print('Approximate 95% confidence interval for the true slope:')
    print(left, right)
```

```python
bootstrap_slope(baby, 1, 0)
```
