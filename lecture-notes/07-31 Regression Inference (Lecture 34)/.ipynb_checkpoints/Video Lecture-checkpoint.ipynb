# Regression Model
The past lecture, we discussed about a model for regression. We had a cloud of data points that appear to follow a linear relationship between `x` and `y`. We treat the data as if there is a real line that the data points follow, and that the data points are not perfect on the line because of errors, variation of random sampling, etc.

## A "Model": Signal + Noise
A model implies that there is a real line that exhibits relationship between `x` and `y`. "if we gain 10 pounds, or babies will gain 10 pounds". 
<img src = 'model.jpg' width = 500/>

The points don't fall exactly on the line, instead we say that the points fall according to "normal distribution with `mean` `0`". This means we expect half of the points fall above the real line, and the other half fall below the real line. 

The important thing is that we think that there is an actual line, and thus we can use a regression line. If instead we think the data follows a curve, then we won't be able to use a linear regression. 

## What We Get to See
In actuality, what we get to see is the following,
<img src = 'see.jpg' width = 400/>

We don't get to see the "real" lines. All we see is the points that are generated from that line. If we draw a regression line from the points, how close is the regression line from the actual line?

Below we have functions that are useful from previous lectures,

```python
def standard_units(any_numbers):
    """ Convert an array of numbers to standard units"""
    # (Value - mean) / SD
    return (any_numbers - np.mean(any_numbers)) / np.std(any_numbers)

def correlation(t, x, y):
    """ Return the correlation coefficient (r) of 2 variables. """
    # Average of the product of the `x` and `y`
    return np.mean(standard_units(t.column(x)) * standard_units(t.column(y)))

def slope(t, x, y):
    """ The slope of the regression line (original units). """
    r = correlation (t, x, y)
    # Slope = r * (SD of y / SD of x)
    return r * np.std(t.column(y)) / np.std(t.column(x))

def intercept(t, x, y):
    """ The intercept of the regression line (original units)"""
    # Average(y) - slope * Average(x)
    return np.mean(t.column(y)) - slope(t, x, y) * np.mean(t.column(x))

def fit(table, x, y):
    """ Return the fitted y value"""
    a = slope(table, x, y)
    b = intercept(table, x, y)
    return a * table.column(x) + b

def plot_residuals(t, x, y):
    """ Plot a scatter diagram and residuals. """
    t.scatter(x, y, fit_line =True)
    actual = t.column(y) # The actual y values
    fitted = fit(t, x, y) # The predicted y values
    residuals = actual - fitted
    print('r:', correlation(t, x, y)) # Print the r
    print('RMSE:', np.mean(residuals ** 2) ** 0.5) # Print the RMSE
    t.select(x).with_column('Residual', residuals).scatter(0, 1) # Plot the residual plot
    
```

```python
def standard_units(any_numbers):
    "Convert any array of numbers to standard units."
    return (any_numbers - np.mean(any_numbers)) / np.std(any_numbers)  

def correlation(t, x, y):
    """Return the correlation coefficient (r) of two variables."""
    return np.mean(standard_units(t.column(x)) * standard_units(t.column(y)))

def slope(t, x, y):
    """The slope of ther regression line (original units)."""
    r = correlation(t, x, y)
    return r * np.std(t.column(y)) / np.std(t.column(x))

def intercept(t, x, y):
    """The intercept of the regression line (original units)."""
    return np.mean(t.column(y)) - slope(t, x, y) * np.mean(t.column(x))

def fit(table, x, y):
    """Return the height of the regression line at each x value."""
    a = slope(table, x, y)
    b = intercept(table, x, y)
    return a * table.column(x) + b

def plot_residuals(t, x, y):
    """Plot a scatter diagram and residuals."""
    t.scatter(x, y, fit_line=True)
    actual = t.column(y)
    fitted = fit(t, x, y)
    residuals = actual - fitted
    print('r:', correlation(t, x, y))
    print('RMSE:', np.mean(residuals**2)**0.5)
    t.select(x).with_column('Residual', residuals).scatter(0, 1)
```

And below we have the plots of data points, actual and residuals

```python
def draw_and_compare(sample_size, true_slope=2, true_int=-5):
    x = np.random.normal(50, 5, sample_size)
    xlims = np.array([np.min(x), np.max(x)])
    errors = np.random.normal(0, 6, sample_size)
    y = (true_slope * x + true_int) + errors
    sample = Table().with_columns('x', x, 'y', y)

    sample.scatter(0, 1)
    plt.plot(xlims, true_slope*xlims + true_int, lw=2, color='green')
    plt.title('True Line, and Points Created')
    plt.show()

    sample.scatter(0, 1)
    plt.title('What We Get to See')
    plt.show()

    sample.scatter(0, 1, fit_line=True)
    plt.title('Regression Line: Estimate of True Line')
    plt.show()

    sample.scatter(0, 1, fit_line=True)
    plt.plot(xlims, true_slope*xlims + true_int, lw=2, color='green')
    plt.title("Regression Line and True Line")
    plt.show()
    
_ = interact(draw_and_compare, sample_size=make_array(10, 100, 400))
```

Above, as we increase the sample size, the regression line (blue line) will become closer to the true line (green line).

# The True Slope
How close is the regression line slope to the actual line slope?

## Confidence Interval for True Slope
Suppose we have a population and we are trying to estimate the parameter (e.g. `mean` baby weights). 

We take a sample, look at the `mean` baby weight of the sample


1. Bootstrap the scatter plot
    * We have data points. We can just select at random from those data points again with replacement. This will give us variation to the scatter plot
    
    
2. Find the slope of the regression line through the bootstrapped plot
    * Take a sample
    * Bootstrap them to obtain a resample 
    * Find the regression line from the resample 


Again, we assume that:
* The original sample is similar to the population
* The resamples look like the original sample


3. Repeat to obtain a bunch of different slopes


4. Draw the empirical histogram of all the generated slopes
    * Make sure the histogram of the generated slopes is roughly bell shaped so that it confirms that it meets the bootstrap criteria


5. Obtain the middle 95% interval

From the steps above, we can obtain an approximate 95% confidence interval for the slope of the true line. It gives us **a range of possible slopes that the actual slope might be in**.

Now we can do multiple resamples and construct the histogram!

```python
slopes = [] # An array that contains all the generated slope
for i in np.arange(5000):
    resample = baby.sample()
    resample_slope = slope(resample, 1, 0)
    slopes.append(resample_slope)
    
Table().with_column('Bootstrap Slopes', slopes).hist(bins = 20)
```

This means that if we increase the gestational days by 1, we expect the baby weights to increase by 0.35 to 0.60 ounces. 

If we calculate the 95% interval,

```python
left = percentile(2.5, slopes)
right = percentile(97.5, slopes)
[left, right]
```

Then we are 95% confident that the real population slope lies between the 2 numbers above.

Below is a function that combines the steps that we did above,

```python
def bootstrap_slope(table, x, y, repetitions = 5000):
    """ 
    table: a table that contains a sample (not population!)
    x and y: labels of data column that we want to analyze
    """
    
    # Bootstrap Resampling
    slopes = []
    for i in np.arange(5000):
        resample = table.sample()
        slopes.append(slope(resample, x, y))
        
    # Find the endpoints of the 95% confidence interval for the true slope
    left = percentile(2.5, slopes)
    right = percentile(97.5, slopes)
    
    # Slope of the regression line from the original sample
    observed_slope = slope(table, x, y)
    
    # Display results
    Table().with_column('Bootstrap Slopes', slopes).hist(bins = 20)
    plt.plot([left, right], [0, 0], color = 'yellow', lw = 8);
    print('Slope of regression line:')
    print(observed_slope)
    print('')
    print('Approximate 95% confidence interval for the true slope:')
    print(left, right)
```

```python
bootstrap_slope(baby, 1, 0)
```

## Rain on the Regression Parade
<img src = 'what_if.jpg' width = 500/>

What if it just happened that we have a sample that has a slightly zero slope or slightly positive slope? What if the the true line is actually a flat, that there's no relationship between the `x` and `y` axis? 

## Test Whether There Really is a Slope
If there's a relationship between `x` and `y`, when we take multiple samples, we shouldn't get a `0` slope. On the other hand, if there is no relationship between `x` and `y`, then we should get sometimes a little positive slope, sometimes a little negative slope, sometimes `0`. 

If want to test whether there really is a slope (a relationship between `x` and `y`), then we can construct that as a hypothesis test.

1. `Null` hypothesis: The slope of the true line is `0` 
    * There is no correlation between `x` and `y`
2. `Alternative`: The slope is not `0`

The methods to do the test is similar to what we have done so far:
1. Construct a bootstrap confidence interval for the true slope
2. If the interval doesn't contain `0`, then reject the null hypothesis
3. If the interval does contain `0`, then there isn't enough evidence to reject the null hypothesis
    * From bootstrapping, if we obtain some positive slopes and some negative slopes, this means our confidence interval contains `0`
    * This means there is no relationship between `x` and `y` 
    
    
Below is the 95% confidence interval of the slope for the relationship between `Maternal Age` and `Gestational Days`.

```python
bootstrap_slope(baby, 2, 1)
```

Above is an example of a case where it's difficult to make a conclusion since the data is close to the cutoff. The confidence interval appears to barely include `0`, which means we should say that we can't reject the null. However, we would be able to make conclusion easier if we have a smaller or greater interval.

Here we see that the bootstrap method of taking slopes lets us answers questions about the data. Instead of just a single prediction line, we can obtain multiple predictions line. With this, we can make a conclusion that the true slope lies within a range instead of saying that the true slope is a certain value. This is a more accurate conclusion compared to relying on one sample only, since it is possible for that one sample to be off from the population. 

# Prediction Variability
Just as how we can bootstrap the slope, we can also bootstrap the prediction and obtain a range of predictions for a given `x` value. 

## Regression Prediction
If the data come from the regression model (half of the time above, half of the time below), then
1. The regression line is close ot the true line
2. Given a value `x`, `y` can be predicted by finding the point on the regression line at that `x`
3. Bootstrap the scatter plot
4. Get a new prediction using the regression line that goes through the resampled plots

Repeat steps `3` and `4` multiple times. From the repetition, take the middle 95% of the predictions. That way, we have a 95% confidence interval of predictions of `y` for the given `x`. Thus, we can predict not only slope, but also the value `y`! 

## Prediction Inference
Below we have a plot of `Gestational Days` vs. `Birth Weight`, along with the regression line (represented as the blue line) and the `Birth Weight` prediction when the `Gestational Days` is `300`.

```python
x = 300
a = slope(baby, 1, 0)
b = intercept(baby, 1, 0)
fitted_y = a * x + b # fitted_y is the predicted y
baby.scatter(1, 0, fit_line=True) # Construct the scatter plot with the regression line
plt.plot([x, x], [40, fitted_y], color='gold'); # Plot the vertical yellow line
print('The predicted y is:')
print(fitted_y)
```

Above, when we have a `Gestational Days` of `300`, it seems that the predicted `Birth Weight` is `129.21`. 

Below we have a simulation of regression lines and the predicted `y` values with varying number of predictions,

```python
lines = Table(['slope','intercept', 'at 291', 'at 300', 'at 309'])

for i in range(100):
    resample = baby.sample()
    a = slope(resample, 1, 0)
    b = intercept(resample, 1, 0)
    lines.append([a, b, a * 291 + b, a * 300 + b, a * 309 + b])

def plot_lines(n):
    '''Plots n prediction lines'''
    for i in np.arange(n):
        line = lines.row(i)
        plt.plot([291, 309], [line.item('at 291'), line.item('at 309')], lw=1)
        plt.scatter(300, line.item('at 300'), s=30)
    plt.show()
_ = interact(plot_lines, n=make_array(1, 10, 50, 100))
```

Above, we see that the prediction is ranging between `127` to `132`. This way, even though we only have one original sample, instead of concluding that the sample predicts `129.21`, we can conclude that based on the variability of the sample, the **real population value** lies within a range of `127` to `132`. 

For convenience, we can make a function `bootstrap_prediction` that does the bootstrap and eventually results in a 95% confidence interval of the `y` predictions,

```python
def bootstrap_prediction(table, x, y, new_x, repetitions = 5000):
    """
    table: a table that contains the original sample
    x and y: column labels
    new_x: the `x` value that we want to predict
    """
    
    # Bootstrap
    predictions = []
    for i in np.arange(repetitions):
        resample = table.sample() # Take resamples
        a = slope(resample, x, y)
        b = intercept(resample, x, y)
        fitted_y = a * new_x + b # Predicted y
        predictions.append(fitted_y) # Append the predicted y to the array 'predictions'
        
    # 95% Confidence Interval
    left = percentile(2.5, predictions)
    right = percentile(97.5, predictions)
    
    # Display results
    Table().with_column('Prediction', predictions).hist(bins=20)
    plt.xlabel('predictions at x='+str(new_x))
    plt.plot([left, right], [0, 0], color='yellow', lw=8);
    print('Approximate 95%-confidence interval of regression height:')
    print(left, right, '(width =', right - left, ')')
```

```python
bootstrap_prediction(baby, 'Gestational Days', 'Birth Weight', 300)
```

And indeed! We obtain a prediction range of `127` to `131`! The **width** of the interval represents the uncertainty of the prediction. 

## Predictions at Different Values of `x`
Since `y` is correlated with `x`, the predicted `y` values depend on the value of `x` 

The **width of the prediction interval** also **depends on** `x`. Typically, interval are wider for values of `x` that are further away from the `mean` of `x` because the further we are from the `mean` of `x`, the less data points that we can use. Wider interval means we are **less certain** of the prediction. 

Recall we had the prediction of `129.21` from the original sample,

```python
fitted_y
```

And from the bootstrap method above, we obtain an interval between `127` to `131`.

The width of the interval is important! If we look back to the scatter plot,

```python
baby.scatter(1, 0, fit_line=True) # Construct the scatter plot with the regression line
plt.plot([x, x], [40, fitted_y], color='gold'); # Plot the vertical yellow line
```

As we resample, we'll obtain lines that look slightly different. On average, the lines would go through the center of the data (middle of the cloud) with slightly varying slopes and intercepts. 

We made a prediction with `Gestatonal Days` of `300`. What about `250` days? What about `350` days?

Based on the scatter plot, it would be difficult to make a prediction for `350` or somewehere below `250` `Gestational Days`. This is because we don't have that much data around that `x` values. Most data points are present in the range of `260` to `300` Gestational Days. If we try to make a prediction on `350` Gestational Days,

```python
bootstrap_prediction(baby, 'Gestational Days', 'Birth Weight', 350)
```

Notice that the width of the interval increases! As we get further from the concentration of data, the confidence interval becomes wider, indicationg that we become less certain with the prediction. On the other hand, if we try to make a prediction in the area where the data is concentrated,

```python
bootstrap_prediction(baby, 'Gestational Days', 'Birth Weight', 280)
```

Above, the width of the interval decreases!

## Recap
Linear regression only gives us one estimate (one sample, one line, one slope, one estimation) for every `x` value. Bootstrapping that sample gives us a range of slopes and predictions. 

With bootstrap, from one sample we can draw confidence interval for both slope and prediction.
