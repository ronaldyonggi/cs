1. Numerical data, where each value is a number from a numerical scale

# Binning
It appears that a plain bar chart is not exactly the best method to visualize numerical data. If we have older movies in the table, the bar that corresponds to that movie would be located far down of the bar chart. 

The problem with the bar chart above was that it compared individual years. We are interested with the overall trend, whether there were younger movies or older movies. We can use `binning` to lump individual years together and see the overall trend.

## Binning Numerical Values
When we bin numerical values, we take individual values and group them in ranges (or `bins`). For example, "this bins goes from 10 to 20..."

1. Typically the range for bins is inclusive on the left side (or the lower bounds)
2. The upper bound is the lower bound of the next bin

<img src = 'bins.jpg' width = 600/>
Above, 185 falls to 185-190 bin, but 190 would fall to the next bin.

If we take these bins and create a "bar chart" out of it, we will get a `histogram`.

# Histogram
Histogram is a chart that displays the distribution of numerical values using bins. It's basically a bar chart that uses bins.

If we take a look at the `Age` count once again,

```python
aged.group('Age')
```

We can use the `bin` method on the table above,

```python
aged.bin('Age').show()
```

In the table above, there are:
1. 41 counts in bins 2 to 11.4 (not including 11.4). 
2. 40 counts in bins 11.4 to 20.8 (not including 20.8)

...and so on. Notice that on the last row, we have a row with bin 96 and count 0. This is because 96 is supposed to be the upper bound of the last bin.

Typically, we want to create our own custom bins. We can specify our own bins!

```python
aged.bin('Age', bins = make_array(0, 20, 40, 60, 80, 100))
```

Note that the last bins does not mean the value and above! For example, the last bin 80-100 above does not include movies who are above 100 years old.

Here are other methods of creating the same bins:

```python
# Using a list
aged.bin('Age', bins = [0, 20, 40, 60, 80, 100])
```

```python
# Using np.arange
aged.bin('Age', bins = np.arange(0, 101, 20))
# Recall np.arange excludes the endpoint, so the last value out of the np.arange would be 100!
```

Now to create a histogram, we use the `hist` method.

```python
aged.hist('Age')
```

When Python created the histogram above, it used a standard bins of 10. We can adjust the bins as we created previously!

```python
aged.hist('Age', bins = np.arange(0, 101, 20))
```

From the histogram above, we can see that there are younger movies than older movies. 

Now notice the y-axis in the histogram above uses **Percent per unit**. If don't want to use this y-axis system, we can add an additional argument `normed = False`.

```python
aged.hist('Age', bins = np.arange(0, 101, 20), normed=False)
```

# The Density Scale
Why did histograms used the strange "percent per unit" y-axis earlier?

## Histogram Axes
By default, histogram uses **density scale** (`normed = True`). This means the area of the charts sums to 100%.
For example:
1. On the horizontal (x-axis) we have age (e.g. year)
2. On the vertical (y-axis) we have a **percent per unit**
3. The area of a bar is a percentage of the whole

When we multiply the age with **percent per unit**, the final result will be percentage. If we sum up all the bars percentage, they will add up to 100%.

Why this feature is useful?
## Demo
If we change the histogram's bin to display a bar every 5 years, the histogram would display unnecessary details. 

```python
aged.hist('Age', bins = np.arange(0, 101, 5), normed=False)
```

We can smooth out the right end of the histogram (80 to 100) by changing the bins,

```python
aged.hist('Age', bins=[0,   5,  10,  15,  20,  25,  30,
                       35,  40, 45, 50, 55, 60, 65, 70, 75, 80, 100], normed=False)
```

Below we smooth out the range 60-80 as well,

```python
aged.hist('Age', bins=[0,   5,  10,  15,  20,  25,  30,
                       35,  40, 45, 50, 55, 60, 80, 100], normed=False)
```

And below we smooth out the range 40-60,

```python
aged.hist('Age', bins=[0,   5,  10,  15,  20,  25,  30,
                       35,  40, 60, 80, 100], normed=False)
```

Above, we got a tall bar on the range 40-60! This histogram does not resemble our original historgram! What happened?

If we count the bin,

```python
aged.bin('Age', bins=[0,   5,  10,  15,  20,  25,  30,
                       35,  40, 60, 80, 100], normed=False)
```

Above, we can see that there are 44 movies that are within the range 40-60! This is reasonable because as we increased the size of the bin, more movies fall into the bin. 

This the downside of using `count` or `non-density` y-axis scale: **it violates the area principle** (recall the car accident example). The histogram looks misleading and it shows as if there were more movies of 40-60 years old compared to younger movies. The more we expand the bins, the more misleading the histogram becomes.

```python
aged.hist('Age', bins=[0,   5,  10,  15,  20,  25,  30,
                       35,  40, 80, 100], normed=False)
```

With density scale, instead of making the bars as tall as the number of movies, it makes sure each bar's area corresponds to the proportion (%) of the movies that fall to each bin. If we change the histogram above to `normed = True`,

```python
aged.hist('Age', bins=[0,   5,  10,  15,  20,  25,  30,
                       35,  40, 80, 100], normed=True)
```

The resulting histogram looks more reasonable, resembles the original histogram. 

## How to Calculate Height
We can't use the number of counts to define height. So how do we calculate the height?

The area of a bar corresponds to the proportion of movies that fall into that bar. 

Let's say the \[20, 40) bin contains 59 out of 200 movies.
1. 59 out of 200 is `29.5\%`
    * We want the `width $\times$ height` of the bar to be equal `29.5`
2. The bin (the bar) width is `40-20` = `20` years wide

**Height of bar = $\frac{29.5 \%}{20 years}$** = `1.475` percent per year

If we don't count the actual number of movies, how do we interpret the strange density scale percent per year?

One interpretation is that the **height measures density**.

## Height Measures Density
Height  = `% of bin` / `width of bin`

The height measures the percent of data in the bin **relative to the amount of space in the bin**.

If we originally have 30 movies in a bin and we stretch out the bin twice as long, the bin would become half as tall. This is how the density scale avoids the pitfall of misleading histogram.

The height measures crowdedness, or **density**.

## Area Measures Percent
This is just a rearrangement of the equation.

Area = `% of bin` = height $\times$ width of bin

1. If we want to know, "how many individuals are in the bin?" or "how many movies are in the bin?"
    * Compute the `area`
    * Then multiply the % area with the total number of individuals or movies
2. "How crowded is the bin?" 
    * Use `height`
    * Recall the `height` represents density
    
Back to the `aged` histogram, if we look the histogram below

```python
aged.hist('Age', bins=[0,   5,  10,  15,  20,  25,  30,
                       35,  40, 100], normed=True)
```

```python
aged.bin('Age', bins=[0,   5,  10,  15,  20,  25,  30,
                       35,  40, 100], normed=False)
```

The bins at 40-100 contains 65 movies, which makes up about 32% of the whole 200 movies. The area of the bar from bins 40 to 100 should make up about 32%. 

#### If the bin 0-10 is combined, what should be the height of the bin?
1. Width is 10 years.
2. There are 33 out of 200 movies, so 16.5% is the area of the bin
3. Thus the height would be `16.5 % / 10 years` = `1.65` percent per unit

We ca double check if this is correct by constructing the histogram below,

```python
aged.hist('Age', bins=[0, 10,  15,  20,  25,  30,
                       35,  40, 100], normed=True)
```

And we can see above that the height of the first bar is roughly 1.6.

# Recap: Chart Types

## Bar Chart vs. Histogram

#### Bar Chart
1. Bar chart has a categorical axis and a numerical axis
2. Bars have arbitrary (but equal) widths and spacing
3. If we use bar charts for making distribution, the height or length of the bars are proportional to the % of individuals that fall to that bar or category

#### Histogram
1. The horizontal axis is numerical, and thus there's no gaps or spaces in between bars
2. Height measures density, areas are proportional to the percent of individuals

# Overlaid Graphs
Let's say we want to compare between 2 distributions or populations.

Below we have a dataset from Francis Galton, an English scientist. He observe the height of people and compared them to the height of their parents. The dataset below are in inches.

```python
heights = Table.read_table('galton_subset.csv')
# Rearrang the table so 'son' appears first
heights = heights.select('son', 'father', 'mother')
heights
```

Below is the histogram of the `son`,

```python
heights.hist('son')
```

While below is the histogram of the `father`,

```python
heights.hist('father')
```

It is difficult to compare the 2 histograms above because of the different axes and that we have to scroll up and down. In this case, we can create an overlay histogram,

```python
heights.hist('son', 'father')
```

Here we can compare the 2 distributions easily. We can see that the sons are in general taller than their fathers.

We can compare all 3 distributions as well,

```python
heights.hist('son', 'father', 'mother')
```

We can add a unit to the histogram by adding the argument `unit = `.

```python
heights.hist('son', 'father', 'mother', unit = 'inch')
```

We can change the bins as well,

```python
heights.hist('son', 'father', 'mother', 
             bins = np.arange(55, 81, 1),
             unit = 'inch')
```

If we want to change the x-axis label, we can do the following,

```python
heights.hist('son', 'father', 'mother', 
             bins = np.arange(55, 81, 1),
             unit = 'inch')
plt.xlabel('Height (inches)')
```

Now we have `Height (inches)` as the x-axis (corresponds to `plt.xlabel`) and Percent per `inch` (corresponds to the `unit` argument) as the y-axis

This histogram helps us compare overall distributions, but it can't compare individual heights. If we want to compare individual heights, we use `scatter` plot.

```python
heights.scatter('son', 'father')
```

From the `scatter` plot above, we can observe that there appears to be a linear correlation between the height of the son and the height of the father. This observation can't be found with histogram.

One neat shortcut for making histogram is that if we don't give any argument for which column to plot, Python will make a histogram of all the available columns.

```python
heights.hist( 
             bins = np.arange(55, 81, 1),
             unit = 'inch')
plt.xlabel('Height (inches)')
```

We can also overlay scatter plot. The argument given in the scatter plot below specifies which column will be in x-axis.

```python
heights.scatter('son')
```
