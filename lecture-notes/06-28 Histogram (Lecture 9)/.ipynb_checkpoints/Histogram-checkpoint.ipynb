# Binning

## Binning Numerical Values
Binning is counting the number of numerical values that lie within ranges, called **bins**
* Bins are defined by their lower bounds (inclusive)
* The upper bound is the lower bound of the next bin
<img src = 'bin.jpg'/>

## Histogram
Chart to display the distribution of numerical values using bins

```python
# We're going to bin this table of ages!
top_age.group('Age')
```

```python
top_age.bin('Age')
```

How you read the bin above, there are
* 41 data points lie between the range 2 to 11.4
* 40 data points lie between the range 11.40 to 20.8
* and so on...
Notice that there's 1 row omitted. If we decide to show everything,

```python
top_age.bin('Age').show()
```

You'll notice that there's 96 with 0 a count. **bin** requires upper and lower bounds and thus, this 96 only serves as an upper bound!
Also notice the strange numbers 11.4, 20.8, etc. By default, **bin** takes data and and divides it into 10 equal sections. However, you can specify your own **bin**.

```python
top_age.bin('Age', bins = make_array(0, 20, 40, 60, 80, 100))
```

As a shorcut, you can use a list directly instead of **make_array*.

```python
top_age.bin('Age', bins = [0, 20, 40, 60, 80, 100])
```

Or you can use np.arange.

```python
top_age.bin('Age', bins =np.arange(0, 101, 20))
```

We can make a histogram using the **hist** method.

```python
top_age.hist('Age')
```

Notice in the chart above that the bins don't align exactly with the x-axis gridlines. Again, this is because Python by default creates bins that may not be exactly as what you wanted. You can always specify the bins by yourself like below:


```python
top_age.hist('Age', bins = np.arange(0, 101, 20))
```

From the histogram above, we can clearly see that there are more younger movie than older movies!
Notice the Y-axis "Percent per unit" that you have never seen before. If you don't want this unit, we can revert it back to 'count'.

```python
top_age.hist('Age', bins = np.arange(0, 101, 20), normed = False)
```

# The Density Scale

## Histogram Axes
By default, **hist** uses a scale (**normed = True**) that ensures the area of the chart sums to 100% (if you multiply the vertical value with the horizontal value and add them up, they'll total to 100%).
* The horizontal axis is a number line (i.e. years)
* The vertical axis is a rate (i.e. percent per year)
* The area of a bar is a percentage of the whole

## Demo
Going back to the previous histogram, we'll analyze a finer version of the histogram.

```python
top_age.hist('Age', bins = np.arange(0, 101, 5), normed = False)
```

This is too much details, too much ups and downs! We can smooth the historgram by specifying the gap of the bins. Here's an example of smoothing the range from 80 to 100:

```python
top_age.hist('Age', bins = [ 0,  5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80, 100], normed = False)
```

Notice above that in the array that we specify as the bins, we jumped straight from 80 to 100! Here's another example that smoothens the bars from 60 to 80:

```python
top_age.hist('Age', bins = [ 0,  5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 80, 100], normed = False)
```

Now we'll smooth out 40-60 range:

```python
top_age.hist('Age', bins = [ 0,  5, 10, 15, 20, 25, 30, 35, 40, 60, 80, 100], normed = False)
```

Now the bar for age range 40-60 appears to be the highest! What happened here?
If we look at the Y-axis, the axis is different compared to the previous histogram. The amount of movie ranged from 40-60 years are clumped into one bar and thus, their amount are summed and became much greater than the other bars. **This is misleading! We don't want this**.

If you clumped the bin even more, the representation would become even worse. See below:

```python
top_age.hist('Age', bins = [ 0,  5, 10, 15, 20, 25, 30, 35, 40,80, 100], normed = False)
```

This is where the **density scale** comes in! The density scale (recall the percent per unit Y-axis earlier) makes sure each bar area corresponds to the actual proportion.

```python
top_age.hist('Age', bins = [ 0,  5, 10, 15, 20, 25, 30, 35, 40,80, 100])
```

Now the charts look more reasonable!

## How to Calculate Bar Height
Recall the previous charts by count with 20 years gap. The [20,40) bin contains 59 out of 200 movies.

* 59 out of 200 is 29.5%
* The bin is 40-20 = 20 years wide

Height of bar = 29.5 percent / 20 years = 1.475 percent per year

### Height Measures Density
Height = % in bin / width of bin

The height measures the % of data in the bin **relative to the amount of space in the bin**. Thus, height measures crowdedness, or **density**.

### Area Measures %
Area = % in bin = height * width of bin

* "How many individuals in bin?" Use **area**
* "How crowded is the bin?" Use **height**

| Bar Chart | Histogram |
| :---------: | :---------:|
| 1 categorical axis| Horizontal axis is numerical, hence to scale with no gaps |
| 1 numerical axis| Height measures density |
| Bars have arbitrary (but equal) widths and spacing| Areas are proportional to the percent of individuals
| For distributions: height (or length) of bars are proportional to the percent of individuals|

All values in a column should be both the same type **and** be comparable to each other in some way

* SEX is a categorical variable even though the values have been given a numerical code.

## Visualizing the Distribution of the Adjusted Receipts
In this section we will draw graphs of the distribution of the numerical variable in the column **Gross (Adjusted)**. For simplicity, let’s create a smaller table that has the information that we need. And since three-digit numbers are easier to work with than nine-digit numbers, let’s measure the Adjusted Gross receipts in millions of dollars. Note **round** is used to retain only two decimal places.

```python
# Notice the first argument in np.round, the top.column array is divided by 1 million!
millions = top.select('Title').with_column('Adjusted Gross', 
                                     np.round(top.column('Gross (Adjusted)')/1e6, 2))
millions
```

## A Histogram
A histogram of a numerical dataset looks very much like a bar chart, though it has some important differences that we will examine in this section. First, let’s just draw a histogram of the adjusted receipts.

The **hist** method generates a histogram of the values in a column. The optional unit argument is used in the labels on the two axes. The histogram shows the distribution of the adjusted gross amounts, in millions of 2016 dollars.

```python
millions.hist('Adjusted Gross', unit="Million Dollars")
```

### The Horizontal Axis
The amounts have been grouped into continuous intervals called **bins**. Although in this dataset no movie grossed an amount that is exactly on the edge between two bins, hist takes into account values that might have been at the edges. So hist has an endpoint convention: **bins include the data at their left endpoint, but not the data at their right endpoint**.

We will use the notation 
$$[a, b)$$ 
for bin that **starts at a and ends at b but doesn’t include b**.

Sometimes, adjustments have to be made in the first or last bin, to ensure that the smallest and largest values of the variable are included. You saw an example of such an adjustment in the Census data studied earlier, where an age of “100” years actually meant “100 years old or older.”

We can see that there are 10 bins (some bars are so low that they are hard to see), and that they all have the same width. We can also see that none of the movies grossed fewer than 300 million dollars; that is because we are considering only the top grossing movies of all time.

It is a little harder to see exactly where the ends of the bins are situated. For example, it is not easy to pinpoint exactly where the value 500 lies on the horizontal axis. So it is hard to judge exactly where one bar ends and the next begins.

The optional argument bins can be used with hist to specify the endpoints of the bins. It must consist of a sequence of numbers that starts with the left end of the first bin and ends with the right end of the last bin. We will start by setting the numbers in bins to be 300, 400, 500, and so on, ending with 2000.

```python
millions.hist('Adjusted Gross', bins=np.arange(300,2001,100), unit="Million Dollars")
```

The horizontal axis of this figure is easier to read. The labels 200, 400, 600, and so on are centered at the corresponding values. The tallest bar is for movies that grossed between 300 million and 400 million dollars.

A very small number of movies grossed 800 million dollars or more. This results in the figure being “skewed to the right,” or, less formally, having “a long right hand tail.” Distributions of variables like income or rent in large populations also often have this kind of shape.

### The Counts in the Bins
The counts of values in the bins can be computed from a table using the bin method, which takes:
* 1st argument: a column label or index and 
* 2nd argument (optional): sequence or number of bins

The result is a **tabular form of a histogram**. 
* The first column lists the left endpoints of the bins (but see the note about the final value, below)
* The second column contains the counts of all values in the Adjusted Gross column that are in the corresponding bin
    * It counts all the Adjusted Gross values that are greater than or equal to the value in bin, but less than the next value in bin

```python
bin_counts = millions.bin('Adjusted Gross', bins=np.arange(300,2001,100))
bin_counts.show()
```

Notice the bin value 2000 in the last row.
* That’s not the left end-point of any bar – it’s the right end point of the last bar 
* By the endpoint convention, the data there are not included
* The corresponding count is recorded as 0
    * Even if there had been movies that made more than $2,000$ million dollars, it would still be recorded as 0
* When either bin or hist is called with a bins argument, the graph only considers values that are in the specified bins.

Once values have been binned, the resulting counts can be used to generate a histogram using the bin_column named argument to specify which column contains the bin lower bounds.

```python
bin_counts.hist('Adjusted Gross count', bin_column='bin', unit='Million Dollars')
```
