# Correlation (Review)
In the previous lecture, we talked about the correlation coefficient `r`. 
1. This `r` measures linear association  
2. It's based on `standard units`. 
3. Possible `r` values are:
\begin{equation*}
-1 \le r \le 1
\end{equation*}

* If `r` = `1`, then the scatter plot is perfect straight line with upwards slope
* If `r` = `-1`, then the scatter plot is perfect straight line with downwards slope
* If `r` = `0`, then we say the `2` variables have no linear association 
    * Be careful with `r` = `0`. `r` only measures linear association. We can have a data that looks like a curve, but that doesn't mean there's no relationship between the 2 variables. It just means that there's no linear relationship.
<img src = 'correlation.jpg' width = 800/>

## Definition of `r`
Recall that **Correlation Coefficient (`r`)** is the **average of product of `x` in standard units and `y` in standard units**. It measures how clustered the scatter is around a straight line.

# Properties of Correlation
## Properties of `r`
`r` is a pure number without units.
* Recall when we convert things to standard units and obtain `z` of `1` or `-1`, this number actually doesn't have any unit. 
* Think of it as calculating `SD`, but when we do the cancellation of units, we'll eventually cancel out all units. 

`r` is not affected by changing units of measurement
* Even if we change inches to yards, or ft to meters, etc., the `r` will remain unchanged

`r` is not affected by switching the horizontal and vertical axes

## Demo
Recall the `hybrid` car table from the previous lecture, and we filtered the table to only show SUVs.

```python
hybrid = Table.read_table('hybrid.csv')
suv = hybrid.where('class', 'SUV')
suv.show(3)
```

We looked at the relationship between `acceleration` and cost,

```python
suv.scatter('acceleration', 'msrp')
```

We also have the functions that we defined in the previous lecture.

```python
def standard_units(any_numbers):
    "Convert any array of numbers to standard units."
    return (any_numbers - np.mean(any_numbers)) / np.std(any_numbers)  

def standardize(t):
    """Return a table in which all columns of t are converted to standard units."""
    t_su = Table()
    for label in t.labels:
        t_su = t_su.with_column(label + ' (su)', standard_units(t.column(label)))
    return t_su

def correlation(t, x, y):
    """Return the correlation coefficient (r) of two variables."""
    return np.mean(standard_units(t.column(x)) * standard_units(t.column(y)))
```

Now recall we looked at the correlation of `acceleration` and cost,

```python
correlation(suv, 'acceleration', 'msrp')
```

Looking back at the plot and matching it with the correlation value, the data points look like a cloud rather than a linear straight line. However, there is indeed a somewhat positive trend! 

Now let's add another column to the SUV table. This new column, `msrp(1k)` is the `msrp` in thousands of dollars.

```python
suv_1k = suv.with_column('msrp (1k)', suv.column('msrp') / 1000)
suv_1k
```

Now we will show that `r` doesn't change even though we change the units of any of the axes. If we plot `acceleration` vs. `msrp (1k)`,

```python
suv_1k.scatter('acceleration', 'msrp (1k)')
```

The data points look exactly the same as the ones in the original plot! However, what if we calculate the `r`?

```python
correlation(suv_1k, 'acceleration', 'msrp (1k)')
```

The `r` value is the same as before! Thus, changing units does not affect `r` since `r` measures the tightness of the linear relationship.

What if we switch the `x` and `y` axes?

```python
suv_1k.scatter('msrp (1k)', 'acceleration')
```

The data points look slighlty different! But if we calculate the correlation `r`,

```python
correlation(suv_1k, 'msrp (1k)', 'acceleration')
```

The `r` stays unchanged!

If the data is perfectly on a straight line, the slope of the line doesn't matter as long as the slope of the line is positive. `r` does not measure how the data is sloped, instead it measures how closely tied to each other the data points are.

## Interpreting `r`
Now we're going to talk about interpreting the `r`.

There are some things we have to be careful for with `r`. Watch out for:

1. Jumping to conclusions about causality
    * Even if `r` is `1` (perfect correlation), doesn't mean that `x` causes `y`. 
    * For example, we look at days a person drink tea vs. days that person alive. That person drinks tea every day. The correlation `r` would be `1`, but that doesn't mean that person is alive to drink tea or that person drinks tea to stay alive. 

2. Non-linearity
    * Visualize the data first
        * If the data looks linear, then calculate `r` 
        * Otherwise, it could be misleading if we still calculate `r` despite being non-linear

3. Outliers
    * `r` is based on `mean` and `SD`, which are easily affected by outliers. 
    
Does this mean we can exclude outlier during calculation? NO. Generally speaking, it's not a good idea to take out outliers at random just because it doesn't make the data look good. 

4. Ecological correlations, based on aggregates or averaged data

```python
def stats_ac(i):
    i = str(i)
    x = 'x' + i
    y = 'y' + i
    print(x + ' mean:', np.round(np.mean(ac.column(x)), 2))
    print(y + ' mean:', np.round(np.mean(ac.column(y)), 2))
    print(x + ' SD:', np.round(np.std(ac.column(x)), 2))
    print(y + ' SD:', np.round(np.std(ac.column(y)), 2))
    print('xy' + i + ' r:', np.round(correlation(ac, 'x' + i, 'y' + i), 2))
```

# Linear Regression
If we recall the 2nd week of the class, we had a prediction example! Recall Galton's effort in trying to predict children's heights based on their parents' height,

```python
galton = Table.read_table('galton.csv')
galton
```

```python
heights = galton.select('midparentHeight', 'childHeight').relabeled('midparentHeight', 'MidParent').relabeled('childHeight', 'Child')
heights.show(4)
```

`MidParent` is the average of the 2 parents (father and mother)'s heights. If we make a scatter plot out of the data above,

```python
heights.scatter(0)
```

From the plot above, if the `MidParent` height is `72`, we can estimate that the `Child`'s height is around 60-75. 

At a glance, it appears that there is a somewhat positive correlation between `MidParent` and `Child`. 

Recall we tried to predict the average height of the child for parents whose heights are 68 inches. We took a slice of data between `67.5` to `68.5` (enclosed within the red lines),

```python
close_to_68 = heights.where('MidParent', are.between(67.5, 68.5))
close_to_68.show(3)
```

Then we took the average of the `y` vaues (`Child`) from the separated data points above,

```python
close_to_68.column('Child').mean()
```

Below is a visual representation of the steps above. We found that if the `Midparent` height is `68`, we can expect the `Child`'s height to be around `66`.

```python
heights.scatter(0)
_ = plt.plot([67.5, 67.5], [50, 85], color='red', lw=2)
_ = plt.plot([68.5, 68.5], [50, 85], color='red', lw=2)
_ = plt.scatter(68, 66.24, color='gold', s=40)
```

Recall we made a function out of the steps above, 

```python
def predict_child(parent):
    # Take the data points where the MidParent Height are `parent` plus minus 0.5
    close_points = heights.where('MidParent', are.between(parent - 0.5, parent + 0.5))
    # Return the average data points only for the `Child` height
    return close_points.column('Child').mean()                       
```

With the function above, if we can predict a child's height based on the `MidParent` height.

```python
predict_child(68)
```

```python
predict_child(74)
```

Now we can use the `apply` method to apply the `predict_child` function to all the `MidParent` value in the `heights` table,

```python
heights_and_predict = heights.with_column(
    'Prediction', heights.apply(predict_child, 'MidParent')
)
heights_and_predict
```

If we plot the `Prediction` above in a scatter plot,

```python
heights_and_predict.scatter(0)
```

The yellow line is the prediction based on the average `y` value of the points that are close to the `x` values. It appears that most of the yellow points fall through somewhere in the middle of the blue data points. 

We can use a correlation to construct a prediction line that doesn't wobble much like the yellow data points! 

## Linear Regression
We convert the `heights` to standard units so that the middle of both `x` and `y` axes are `0`. 

```python
standardize(heights).scatter(0)
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.plot([-4, 4], [-4, 4], color='r', lw=2)
plt.plot([2.5, 2.5], [-4, 4], color='g', lw=2)
```

The red line is a line with a slope `1`. The green line is located at the `x` value that we want to predict. Here we want to predict the `Child` height if the `MidParent` in `(su)` is around `2.5`. 

Does the red line looks like a good prediction line for the points? Probably not. If we look at the points covered with the vertical green line, it appears that if the `MidParent`'s height is around `2.5 (su)`, we predict the `Child` height to be around `1 (su)`.

If we draw the same plot with an additional line, which slope is `r`, 

```python
standardize(heights).scatter(0)
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.plot([-4, 4], [-4, 4], color='r', lw=2)
plt.plot([2.5, 2.5], [-4, 4], color='g', lw=2)

r = correlation(heights, 0, 1)
# Draw a line with the slope of the `r`
plt.plot([-4, 4], [-4*r, 4*r], color='dodgerblue', lw=2)
```

See that the blue line goes through the middle of the points!  It turns out that if we calculate the `r` and use it as a slope, we can obtain a good prediction line! This blue line is called **regression line**. This line is more convenient to use compared to the yellow line that we made out of the data points. 

## Graph of Averages
We looked at a graph of averages. We looked at visualization of `x` and `y` pairs. 

1. Group each `x` with a representative `x` value (rounding)
    * For every single `x` value, take some slice of points around that value


2. Average the corresponding `y` values for each group.


This means for every `x` value, we obtain prediction of one `y` value. If the association between `x` and `y` is linear, then the graph of averages tend to fall on the regression line.


## Regression to the Mean
What we going to do is:
1. Take `x` and `y`
2. Convert to standard units

This regression describes both:
* The deviation of `x` from `0` (the average of `x`s), and
* The deviation of `y` from `0` (the average of `y`s) 

<img src = 'regression.jpg' width = 500/>

In the above equation, we take the calculated `r` and use it as a slope for the `x`. If both `x` and `y` are in standard units, the equation above would be the regression line. 

`r` is always between `-1` and `1`. This means **on average, `y` deviates from `0` less than `x` deviates from `0`**.

Note that this is not true for all points - this is only valid for the **averages** 

Recall that switching `x` and `y` axis does not change the `r`. Below we plot the same plot as below, with swapped `x` and `y` axes.

```python
standardize(heights).scatter(1) #Previously the input was `0`, this time the input is `1`
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.plot([-4, 4], [-4, 4], color='r', lw=2)
plt.plot([2.5, 2.5], [-4, 4], color='g', lw=2)

r = correlation(heights, 0, 1)
# Draw a line with the slope of the `r`
plt.plot([-4, 4], [-4*r, 4*r], color='dodgerblue', lw=2)
```

Notice that the blue lines still goes through most of our points! Thus, `r` and the regression line stays unchanged even if we swap the axes.

Now going back to the previous plot,

```python
standardize(heights).scatter(0)
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.plot([-4, 4], [-4, 4], color='r', lw=2)
plt.plot([2.5, 2.5], [-4, 4], color='g', lw=2)

r = correlation(heights, 0, 1)
# Draw a line with the slope of the `r`
plt.plot([-4, 4], [-4*r, 4*r], color='dodgerblue', lw=2)
```

If we have an `x` value of `2.5`, we can predict using the blue line that `y` would be around `0.8`. If we have an `x` value of `-1`, then the predicted `y` would be around `-0.4` or `-0.5`. However, we are interested in the data with the original units. 

If we have a `MidParent` height of 75 inches, how tall would be the child? 

# Slope and Intercept

## Regression Line Equation
In **original units**, the regression line has this equation:
<img src = 'regression_equation.jpg' width = 700/>


In original units, we are trying to solve for `estimate of y`. If we rearrange the equation above, the line can be expressed by `slope` and `intercept`,

\begin{equation*}
y = slope \times x + intercept
\end{equation*}

The above equation is expressed in original units, not standard units.

Below we calculated the parameters that are useful for the prediction calculation,

```python
parent_mean = np.mean(heights.column('MidParent'))
parent_sd = np.std(heights.column('MidParent'))
child_mean = np.mean(heights.column('Child'))
child_sd = np.std(heights.column('Child'))
print('Parent: mean =', parent_mean, '; SD =', parent_sd)
print(' Child: mean =', child_mean, '; SD =', child_sd)
print('     r:', r)
```

With the parameter above, let's create a function `predict_with_r` predicts the height of the child based on the input of `MidParent` height. 

```python
def predict_with_r(parent):
    """ Return a prediction of the height of a child whose parents have a midparent
    height of mp, using linear regression"""
    # 1. Convert the units from original to standard units
    parent_su = (parent - parent_mean) / parent_sd
    # 2. Obtain the predicted child height in standard units
    child_su = r * parent_su
    # 3. Convert the predicted child height from standard units to original units
    return child_su * child_sd + child_mean
```

The steps within the function is as the following:
1. Convert the input parent height to standard units
2. The predicted `y` value can be calculated by multiplying `r` with the `parent` height in `(su)`
3. Convert the predicted `y` value back to the `Child` height in original units.

```python
predict_with_r(68)
```

```python
predict_with_r(74)
```

Above, see that the predicted `Child` height is lower than the `Parent`'s input height! However, if we give an input height that is below the average `Parent` height, the `Child` height that we predict will be closer to the average of the `Parent` height. Recall the average values below:

```python
print('Parent: mean =', parent_mean, '; SD =', parent_sd)
print(' Child: mean =', child_mean, '; SD =', child_sd)
```

```python
# Below 69.2067
predict_with_r(60)
```

The slope of the line is always less than `1`, which means that the predicted `y` value will always be closer to the average value.

Below we have a plot that compares the prediction that we made early in the lecture with the regression line,

```python
heights_and_predict.scatter(0)
plt.plot([64, 76], [predict_with_r(64), predict_with_r(76)], color='dodgerblue', lw=2)
```
