# Linear Regression (Review)

## Regression Line Equation
In the previous lecture, we calculated the correlation coefficient `r` by the following:
1. Take some data points
2. Convert them to standard units
3. Take the product of the `x` and `y` of each data points in standard units
4. Take the average of the products.

We also found that in standard units, the best fit line called **"regression line"** is:
<img src = 'regression.jpg' width = 500/>

We saw that `r` doesn't change even if we flip the `x` and `y` axes. However, according to the equation above, if we want to predict `x` given `y`, we would have to move `r` to the other side of the equation, resulting in a different `r`.

The $y_{(su)}$ and $x_{(su)}$ above are not typical `x` and `y` values. The equation above doesn't mean the equality $y = r \times x$. Instead, it means that given a value of `x`, we can predict the value of `y`. This way, the slope of the line (`r`) is unchanged same even when the axes are flipped.

## Regression Line Equation
Recall that we also converted the `x` and `y` in standard units back to their original units. 

In original units, the regression line has the equation,
<img src = 'equation.jpg' width = 500/>

If we rearrange the equation above, we can obtain the following,

\begin{equation*}
y = slope \times x + intercept
\end{equation*}

**slope of the regression line**:
\begin{equation*}
r \times \frac{SD(y)}{SD(x)}
\end{equation*}

**intercept of the regression line**:
\begin{equation*}
average(y) - slope \times average(x)
\end{equation*}

From these equations, if `x` and `y` are already in standard units, or if both of them have `mean` = `0` and `SD` = `1`, 
1. The slope of the regression line becomes $r \times \frac{1}{1}$, which is just `r`
2. The intercept becomes $0 - slope \times 0$

And we'll get back the original equation
\begin{equation*}
y = r \times x
\end{equation*}

## Discussion Question
A course has a midterm (average = 70, `SD` = 10) and a very hard final (average = 50, `SD` = 12).

If the scatter diagram comparing midterm & final scores for students has a typical oval shape with `correlation = 0.75`, 

1. What do we expect the average final score would be for students who scored `90` on the midterm?
2. How about `60` on the midterm?

Let's have `x` as the midterm and `y` as the final.

1. `x` has `mean` of 70 and `SD` of 10.
2. `y` has `mean` of 50 and `SD` of 12
3. `r` is 0.75

We can define a function that given a midterm score, predicts the final score,

```python
mid_mean, mid_sd = 70, 10
fin_mean, fin_sd = 50, 12
r = 0.75

def predict(midterm):
    # Convert midterm score to standard unit
    mid_su = (midterm - mid_mean) / mid_sd
    # Calculates the predicted final in standard unit
    final_su = r * mid_su
    # Convert the final in standard unit back to original unit
    return (final_su * fin_sd) + fin_mean
```

```python
predict(90)
```

```python
predict(60)
```

Below we have some functions that we defined in the previous lecture,

```python
def standard_units(any_numbers):
    """ Convert any array of numbers to standard units."""
    return (any_numbers - np.mean(any_numbers)) / np.std(any_numbers)

def correlation(t, x, y):
    """ Return the correlation coefficient (r) of 2 variables.
    t = a table
    x, y = a column label
    """
    return np.mean(standard_units(t.column(x)) * standard_units(t.column(y)) )
```

Now recall the following,

**slope of the regression line**:
\begin{equation*}
r \times \frac{SD(y)}{SD(x)}
\end{equation*}

**intercept of the regression line**:
\begin{equation*}
average(y) - slope \times average(x)
\end{equation*}

We can also create functions in which given a table and column labels `x` and `y`, calculate the slope and the intercept of the regression line!

```python
def slope(t, x, y):
    """ Returns the slope of the regression line in original units"""
    # Calculates r
    r = correlation(t, x ,y)
    # Now calculates slope
    return r * np.std(t.column(y)) / np.std(t.column(x))

def intercept(t, x, y):
    """ Returns the intercept of the regression line in original units"""
    return np.mean(t.column(y)) - slope(t, x, y) * np.mean(t.column(x))
```

Now recall the data that we looked at the very first day of lecture: statistics of the novel *Little Women* by Louisa May Alcott.

```python
little_women = Table.read_table('little_women.csv')
little_women = little_women.move_to_start('Periods')
little_women.show(3)
```

Recall that we analyzed the relationship between the periods and the characters. The more periods the chapter has, the more characters the chapter has as well. If we plot them,

```python
little_women.scatter('Periods', 'Characters')
```

For starters, we can calculate the `r` of the plot above,

```python
correlation(little_women, 'Periods', 'Characters')
```

Above, it seems that the plot has a relatively strong linear relationship! Let's try to make a function that takes in a table `t`, column labels `x` and `y`, and returns a prediction of `y` for each `x` based on the regression line!

```python
def fit(t, x, y):
    """ Return the predicted y-value for each x-value"""
    a = slope(t, x, y)
    b = intercept(t, x, y)
    return a * t.column(x) + b
```

Now we'll calculate the `fit`ted values and add it as a new column to the original table,

```python
little_women.with_column(
    'Fitted', fit(little_women, 'Periods', 'Characters')
)
```

Now let's compare the fitted values with the actual data points!

```python
little_women.with_column(
    'Fitted', fit(little_women, 'Periods', 'Characters')
).scatter('Periods')
```

Above, notice that the fitted values form a straight line! This makes sense since we used linear regression.

# Least Squares
Previously, we see the regression line as a correlation:
1. Take `x` and `y`
2. Take the product out of the `2`
3. Take the `Mean` to obtain the correlation 

Another persective is to look at regression line as a line that minimizes error.

## Error in Estimation

Looking at the previous plot, we might think that the best line goes through most data points. In mathematical terms, the best line is one that minimizes the distance of each point from the line. We can quantify this by analyzing the `error`,

\begin{equation*}
Error = Actual - Estimate
\end{equation*}

Errors can be either positive or negative. 

To measure the rough size of the errors:
1. **Square** the `error`s to eliminate cancellation
2. Take the **mean** of the squared errors
3. Take the square **root** to fix units

The steps above is somewhat similar to when we calculated `SD`. We call this the **Root Mean Square Error** (rmse).

## Demo
Below is a function that takes in a `slope` and an `intercept` and does the following:
1. Constructs a regression line based on the input `slope` and `intercept`
2. Constructs vertical error lines that connect some data points to the regression line.

```python
sample = [[131, 14431], [231, 20558], [392, 40935], [157, 23524]]
def lw_errors(slope, intercept):
    print('Slope:    ', np.round(slope), 'characters per period')
    print('Intercept:', np.round(intercept), 'characters')
    little_women.scatter('Periods', 'Characters')
    xlims = np.array([50, 450])
    plt.plot(xlims, slope * xlims + intercept, lw=2)
    for x, y in sample:
        plt.plot([x, x], [y, slope * x + intercept], color='r', lw=2)
```

```python
lw_errors(slope(little_women, "Periods", 'Characters'), intercept(little_women, 'Periods', 'Characters'))
```

What if we don't know the `slope` and `intercept`? What if we just give an input of random `slope` and `intercept`?

```python
lw_errors(50,10000)
```

Notice that the line doesn't go through all the points. On top of that, some of the error lines became longer before. What if we give a negative slope?

```python
lw_errors(-100, 50000)
```

We see above that the error lines are very long!

How do we quantify these errors? Calculate the `rmse`! Below is a function that takes in a `slope` and `intercept`, and calculates the `rmse` based on the *Little Women* novel,

```python
def lw_rmse(slope, intercept):
    lw_errors(slope, intercept)
    x = little_women.column('Periods')
    y = little_women.column('Characters')
    # Calculate the fitted values
    fitted = slope * x + intercept
    # Calculate the (Actual - Fitted), square it, take the average, then take the square root
    return np.sqrt(np.mean((y - fitted) ** 2))
```

If we try `slope` = 50 and `intercept = 10000`,

```python
lw_rmse(50, 10000)
```

This means on average, we expect the points to be about 4,000 away from the regression line.

What about the `rmse` of a negative slope?

```python
lw_rmse(-100, 50000)
```

And indeed, the `rmse` appears huge! Now what if we input the `slope` and `intercept` based on the definition of regression line?

```python
lw_rmse(slope(little_women, 0, 1), intercept(little_women, 0, 1))
```

The calculated `rmse` is only about 2700! 

From above, it seems to be a good idea to define the best line as the line that gives out the smallest `rmse`. 

It turns out when we obtained a line that gives the least `rmse`, we actually got back the regression line as what we have found all this time using `r`.

\begin{equation*}
y = slope \times x + intercept
\end{equation*}

If we try out various slope and intercepts, eventually we'll find that there's no choice of slope and intercept that gives a better error than the calculate `slope` and `intercept` from the regression line.

## Least Squares Line
The error is defined as the `rmse` value. The goal is to find the line that gives the least error, which means the least `rmse` value. The line with the least `rmse` turns out to be the same regression line that we have calculated using `r`.

If we minimizes `rmse`, we also minimizes `mse` (mean squared error). The line with the minimum `rmse` is also the line with the minimum `mse`. There is no significant difference between the `2`, as the only difference is that in `rmse`, we took the square root. 

There are many names for this line. "Best fit" line, least squares line, regression line. We will see these terms being used interchangibly.

Looking back at the plots,

```python
lw_rmse(slope(little_women, 0, 1), intercept(little_women, 0, 1))
```

For the errors, why do we take the vertical distance? Why not diagonal?

We just happened to choose the vertical distance. The truth is, we can define the error in different ways. However, if we defined erorr with the vertical distance, we'll get back the regression line as before.

Why this is important? We could have just done `r` and be satisfied with it. 

Truth is, in the context of linear regression, the `rmse` might not be something new since we already know how to find the line using `r`.

However, if we look at the line with the perspective of minimizing the error, we actually can use different definition of errors and get back different lines!

## Numerical Optimization
Below we have a function that calculates the `mse`, still based on the *Little Women* text,

```python
def lw_mse(any_slope, any_intercept):
    x = little_women.column(0) # x is the Periods
    y = little_women.column(1) # y is the Characters
    fitted = any_slope * x + any_intercept
    return np.mean((y - fitted) ** 2)
```

Here instead minimizing the squared error, we minimize the normal error. 

```python
lw_mse(90, 4000)
```

If we take the square root of the `mse`, we'll get back the `rmse` as we did earlier!

```python
np.sqrt(lw_mse(90, 4000))
```

Imagine that we can try a bunch of different lines and find the one that results in the least error. If we try to do this by hand, it will take forever! Fortunately Python is good at making a bunch of calculation at once. We can tell Python to find the line that minimizes the error and try a bunch of numbers.

Supposed we have the following function,

```python
def f(x):
    return (x - 3) ** 2 + 1
```

If we try some values to the function above,

```python
[f(1), f(2), f(3), f(4), f(5)]
```

It appears that the function above is a parabola. It goes down, and it gets to the point where it can go any lower; instead it will go up. We can see that the value of `x` that minimizes the function is `3`. 

It turns out there's a function in the `datascience` package called `minimize` that takes in a function `f` and returns the input `x` value that results in the lowest `f(x)` value.

```python
minimize(f)
```

We can use `minimize` for functions that takes more than one input as well. Below, we basically tell Python to find the best slope and intercept that gives the least error. 

```python
minimize(lw_mse)
```

We can do linear regression with this `minimize` function! 

## Non-linear regression
Below we have a dataset of shot put, which consists of the weight of the metal ball the athletes used and how far they managed to throw.

```python
shotput = Table.read_table('shotput.csv')
shotput.show(3)
```

```python
shotput.scatter('Weight Lifted')
```

Turns out in `scatter` method, we can add an argument `fit_line` that adds a fit line to the plot. However, this method doesn't give us the slope nor th intercept.

```python
shotput.scatter(0, fit_line = True)
```

Notice with the plot above, there is a data point on the bottom left that got left out! It might seem that the data points fit on a curve rather than a line. 

Previously, we try to fit the data with a line,
$$ ax + b $$

we tweak `a` and `b` so that we got the line that minimizes the square root error.

This time, we try to find the best quadratic curve,

$$ ax^2 + bx + c$$

We tell Python to tweak `a`, `b`, and `c` until it finds a curve that give the least error. Below is a function that finds such quadratic curve, applied to the shotput dataset.

```python
def shotput_squad_mse(a, b, c):
    x = shotput.column(0)
    y = shotput.column(1)
    fitted = a*x**2 + b * x + c
    return np.mean((y - fitted) ** 2)
```

If we just give input of arbitrary `a`, `b` and `c`, we could obtain a very high `mse` value!

```python
shotput_squad_mse(1, 1, 1)
```

Instead, we can apply the `minimize` function to the function `shotput_squad_mse`,

```python
best = minimize(shotput_squad_mse)
best
```

As a brief aside, since the `minimize` above returns 3 values, we can assign the 3 values to 3 different variables in a single line,

```python
a, b, c = best
```

```python
a
```

```python
b
```

```python
c
```

With this `a`, `b` and `c`, we can make predictions of the distances!

```python
x = shotput.column(0)
#quad stands for quadratic
quad_fitted = a * x**2 + b * x + c
quad_fitted
```

```python
shotput.with_column(
    'Predicted', quad_fitted
)
```

Now let's scatter plot it!

```python
shotput.with_column('Best Quad', quad_fitted).scatter(0)
```

Basically, we told Python to find the best `a`, `b` and `c` combination that gives the best curve or line. Other than the quadratic regression, we can do other non-linear regression with this method (e.g. exponential). 

This is actually the basis of many machine learning techniques. All machine learning techniques have complicated functions that calculate error.

## Numerical Optimization
1. Numerical minimization is approximate but effective
2. Lots of machine learning uses numerical minimization
3. If the function `mse(a, b)` returns the `mse` of estimation using the line "estimate = `ax + b`",
    * Then `minimize(mse)` returns array $[a_0, b_0]$
    * $a_0$ is the slope and $b_0$ is the intercept of the line that minimizes the mse among lines with arbitrary slope $a$ and arbitrary intercept $b$ (that is, among all lines)

Regression line lets us make predictions. To compute regression line, there are 2 things that we can do:

2. Or find the slope and intercept that minimize the `mse`
