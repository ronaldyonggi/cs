# Standard Deviation (Review)
* Standard deviation (SD) measures roughly how far the data are from the average
* SD = root mean square of deviations from average
    * What we did was the reverse of this sentence
    * Take average, deviations, square, mean, square root
* SD has the same units as the data
    * We usually write `average` plus or minus a few SDs

# Chebyshev's Inequality
We can use the SD to state that a value is within a certain range. This method involes some mathematics method called **Chebyshev's Inequality**.

## How Big are Most of the Values?
* **Why use Standard Deviation?**
    * No matter what the shape of the distribution (flat, bell, curved, etc.), most of the data are in the range **average $\pm$ a few SDs**. 
* **Chebyshev's Inequality**
    * No matter what the shape of the distribution, the proportion of values in the range **average $\pm$ a few SDs** is at least $ 1 - \frac{1}{z^2}$

## Chebyshev's Bound

| Range | Proportion |
| --- | ---- |
| Average $\pm$ 2 SDs | at least $ 1 - \frac{1}{4}$ (75%)|
| Average $\pm$ 3 SDs | at least $ 1 - \frac{1}{9}$ (88.888...%)|
| Average $\pm$ 4 SDs | at least $ 1 - \frac{1}{16}$ (93.75%)|
| Average $\pm$ 5 SDs | at least $ 1 - \frac{1}{25}$ (96%)|

Read: If we go 2 SDs away, we're going to obtain 75% of the data regardless of the shape of distribution

Now let's go back to the `births` table,

```python
births.show(3)
```

In `Maternal Age`, how many data points fall within 2 SDs?

First of all, get the `average` and `SD`

```python
avg = np.mean(births.column('Maternal Age'))
sd = np.std(births.column('Maternal Age'))
avg, sd
```

Now use the table method `where` and `are.between`.

```python
births.where(2, are.between(avg - 2*sd, avg + 2*sd))
```

Now we'll calculate the percentage of the data above with respect to the full data.

```python
births.where(2, are.between(avg - 2*sd, avg + 2*sd)).num_rows / births.num_rows
```

We have about 95% of the data!

This time, we're going to analyze all the columns. 

```python
births.drop('Maternal Smoker').hist(overlay = False)
```

Below we have a code that calculates the amount of data that we'll obtain for each column for varying SDs.

```python
for i in np.arange(5): # Excluding the `Maternal Smoker`
    values = births.drop('Maternal Smoker').column(i)
    avg = np.mean(values)
    sd = np.std(values)
    print('Below is the proportion of data covered for column', i)
    for z in np.arange(1, 4): # Varying SD between 1 to 3
        within_sd = births.where(i, are.between(avg - z*sd, avg + z*sd))
        print('Average +/-', z, 'SD:', within_sd.num_rows / births.num_rows)
    print() #Prints an empty space between results from each column
```

Oddly enough, the 2 SDs results are roughly around 0.93-0.97, much higher than the theoretical 75%!

# Standard Units

```python
births.hist('Birth Weight')
births.hist('Maternal Age')
```

It was difficult to compare the histograms above because they have different units. We can actually change the units using `Standard Units`.

* Instead of saying "30 eggs away" or "5 km away" from the average, here we say **"2 SDs above/below average"**
* Convert values to the `z` unit. 
    * $ z = \frac{value - mean}{SD} $
        * Negative z : value below average
        * Positive z : value above average
        * z = 0      : value equal to average
* When values are in standard units, average = 0, SD = 1
* By Chebyshev, most values of z are between -5 and 5

## Demo
Going back to our `births` table,

```python
births.show(3)
```

```python
# Take the "Maternal Age" values
values = births.column('Maternal Age')
avg = np.mean(values)
sd = np.std(values)
avg, sd
```

```python
# Calculate z for every value
z = (values - avg) / sd
z
```

```python
t = Table().with_column(
'Values', values,
'Values converted to standard unit z', z)
t
```

```python
t.hist(1)
t.hist(0)
```

Notice that the histogram shape of the converted values look similar to the original data!

In the converted unit histogram, x-axis 1 means 1 SD away to the positive direction while -1 means 1 SD away to the negative.

# The Normal Distribution
In the class, we've gone through examples of data manipulation and simulation. We have seen histograms that look like a bell curve, and it is quite strange that we see bell curve histograms so often. Turns out this shape is called `normal distribution`.

## The SD and the Histogram
* Usually, it's difficult to estimate the SD by looking at a histogram
* But if the histogram has a bell shape, it is easier to estimate the SD

## Demo

```python
# Below is a histogram of the "Maternal Height"
births.hist("Maternal Height", bins = np.arange(57.5, 72, 1), unit = 'inch')
```

We can see above that the histogram of `Maternal Height` looks like a bell curve. At a glance, the SD is where the inflection points of the curve is located.

By inflection points, imagine an upward curve from 58 inch to 64 inch. The curves rises and then reaches a plateau at 64 before it falls down. During the increase, there should be a point where the slope started decreasing (approximately around 61). The inflection point on the other side should be around the same height but on the right side (approximately around 67).

```python
np.std(births.column('Maternal Height'))
```

And indeed! When we calculated it, the SD is around 3!
