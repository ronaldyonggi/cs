# Orders of Growth

The orders of growth of the resources used by a function or program is a method that computer scientists use to categorize how much resources are used without having to be precise about the details.

This is a method for establishing a bound on the resources used by a function that scales by the 'size' of a problem. Instead of picking up a particular input to a function, we say that the function might take any input and the amount of time or space that's needed scales with the input size. 

We need a free variable to describe the size of a problem, and this is often just the input to the function. Other than size, sometimes it can be the length of the input list or some other characteristics of the input.

`n` describes the size of the problem we are trying to solve. 

The function `R(n)` is a measurement of resource used (time or space).
* For example, the time or memory it takes to execute a program

Often times `R(n)` is a complicated function of `n`. Depending on which branch of the `if` statements followed, some inputs require slightly less or more time.

The goal of orders of growth is to bound `R(n)` to a simpler function. An equation that we'll often see is:

$$ R(n) = \Theta(f(n))$$

1. `R(n)` is the true resources used for an input of size `n`
2. `f(n)` is a simple function of `n` (for example, taking the square root of `n`)
3. $\Theta$ establishes a relationship between the equation on the left and the equation on the right

The equation implies that there are positive constants $k_1$ and $k_2$ such that,

$$ k_1 \times f(n) \le R(n) \le k_2 \times f(n) $$

For the equation $ R(n) = \Theta(f(n))$ to be true, the equation above must hold for all `n` larger than some minimum `m`. We can pick any minimum we want and as long as the equation above holds for all larger values.

When we write the equation above, we can think of it as writing 2 equations on the same time:
1. $k_1 \times f(n) \le R(n)$ -> The resources used is bounded below by a function of `n`
2. $R(n) \le k_2 \times f(n)$ -> The resources used is bounded above by the same function of `n`, but with a greater constant

Let's return to the `factors` problem and apply the reasoning!

## Order of Growth of Counting Factors

Recall when defining the `factors(n)` function, we have 2 strategies:

1. Slow: test each `k` from 1 to `n`
2. Fast: test each `k` from 1 to $\sqrt{n}$. For every `k`, $\frac{n}{k}$ is also a factor.

Recall `n` is the input number that we pass in to count factors. 

| Strategy | Time | Meaning | Space | Meaning |
| --- | --- | --- | --- | --- |
| Slow | $\Theta (n)$ | It takes time proportional to the number `n`| $\Theta(1)$ | The amount of space required to compute `factors` is fixed <br> since we have a fixed number of names in the function implementation |
| Fast | $\Theta \text{(} \sqrt {n} \text{)} $ | Grows at a slower pace than above | $\Theta(1)$| Same as above |

Note that the assumption of constant space $\Theta(1)$ assumes that integers occupy a fixed amount of space no matter how large they are.

In practicality, when an integer becomes exceedingly large, they start taking more space. However, we're not going to go in depth to that in this lecture. 

Instead, we'll focus on the difference in order of growth of time for computing `factors`.

Below we have the implementation of `divides` and the 2 `factors` implementation,

```python
from math import sqrt

def divides(k, n):
    """Return whether k evenly divides n."""
    return n % k == 0

def factors(n):
    """Count the positive integers that evenly divide n.

    >>> factors(576)
    21
    """
    total = 0
    for k in range(1, n+1):
        if divides(k, n):
            total += 1
    return total

def factors_fast(n):
    """Count the positive integers that evenly divide n.

    >>> factors_fast(576)
    21
    """
    sqrt_n = sqrt(n)
    k, total = 1, 0
    while k < sqrt_n:
        if divides(k, n):
            total += 2
        k += 1
    if k * k == n:
        total += 1
    return total
```

And below is a code that computes the time it takes to execute the functions above. `timeit` is a built-in module that allows us to `repeat`edly call a function multiple times and record the time it takes to execute the function. 

```python
%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('ggplot')

# The stuff above allows us to draw graphs

from timeit import repeat # timeit is a built-in module that allows us to repeatedly call
# a function multiple times and record the time it takes to execute the function
from numpy import median, percentile # Allows us to compute median and percentile

# name = the function name
# xs = bunch of different arguments to pass


def plot_times(name, xs, n = 25, order = None):
    f = lambda x: name + '(' + str(x) + ')' # Constructs a function call as a string
    g = globals()
    
    samples = []
    for _ in range(n):
        # Repeatedly call the f function 25 times
        times = lambda x: repeat(f(x), globals = g, number = 1, repeat = n)
        samples.append([median(times(x)) for x in xs])
    ys = [10e6 * median(sample) for sample in zip(*samples)]
    
    plots.figure(figsize = (8, 8))
    plots.plot(xs, ys)
    
    if order:
        slopes = [y / order(x) for (x, y) in zip(xs, ys)]
        for slope in (percentile(slopes, 0.2), percentile(slopes, 99.8)):
            plots.plot(xs, [slope * order(x) for x in xs], linewidth = 3)
```

```python
%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('ggplot')

# The stuff above allows us to draw graphs

from timeit import repeat #timeit is a built-in module that allows us to repeatedly call 
# a function multiple times and record the time it takes to execute a function
from numpy import median, percentile # Allows us to compute median and percentile

# name = the function name
# xs = bunch of different arguments to pass in

def plot_times(name, xs, order=None, n=25, pct=0.1):
    f = lambda x: name + '(' + str(x) + ')' # Constructs a function call as a string
    g = globals()
    
    samples = []
    for _ in range(n):
        # Repeatedly calls the f function multiple times
        times = lambda x: repeat(f(x), globals=g, number=1, repeat=n)
        samples.append([median(times(x)) for x in xs])
    ys = [10e6 * median(sample) for sample in zip(*samples)]
    
    plots.figure(figsize=(8, 8))
    plots.plot(xs, ys)
    
    if order:
        slopes = [y / order(x) for (x, y) in zip(xs, ys)]
        for slope in (percentile(slopes, pct), percentile(slopes, 100-pct)):
            plots.plot(xs, [slope * order(x) for x in xs], linewidth=3)
```

```python
args = range(20, 1000)
```

```python
plot_times('factors', args)
```

Above, the horizontal axis is the varying `n`, while the vertical axis is the time in microseconds. 

As we can see, the `factor` is not a simple, linear function (observe the jagged pattern). For example, prime numbers only need 1 and the number itself, whereas a number with many factors has to add many times. 

By contrast, if we plot `factor_fast`,

```python
plot_times('factors_fast', args)
```

It doesn't look like a straight line. Notice the range on vertical axis, which are significantly lower than that of `factor`. 


```python
plot_times('factors', args, lambda x: x)
```

Above, Python looks for constants `k1` and `k2` that determines the slope of the 2 blue lines, which represent the upper bound and the lower bound in order of growth equation. Above is what the lines would look like for `factors`. Notice that we pass in `lambda x:x`, which is a linear function.

Meanwhile, for `factors_fast` function,

```python
plot_times('factors_fast', args, sqrt)
```

And above, we pass in the `sqrt` function to determine the bounds. 

This is the definition of order of growth: what shape of function do we use for upper and lower bounds so that it sandwiches the actual running time (it's also applicable to space).
